## Explicación del código

El siguiente código simula datos de un grupo de personas (estudiantes universitarios), genera estadísticas descriptivas y visualiza la distribución de sus edades y estaturas mediante histogramas. A continuación se explica cada parte:

### 1. Importación de librerías

- **numpy**: para generar números aleatorios con una distribución normal.
- **pandas**: para organizar y manipular los datos en formato tabular.
- **matplotlib.pyplot**: para crear gráficos y visualizaciones.

### 2. Configuración general

- Se fija la semilla aleatoria con `np.random.seed(42)` para garantizar que los resultados sean reproducibles.
- Se define `n = 1000`, es decir, se simulan datos para 1000 personas.

### 3. Generación de datos simulados

- **Edades**: Se generan 1000 edades siguiendo una distribución normal con media 21 años y desviación estándar de 1.5. Los valores se redondean y convierten a enteros para simular edades reales.
- **Estaturas**: Se generan 1000 estaturas siguiendo una distribución normal con media 175 cm y desviación estándar de 8 cm. Los valores se redondean a un decimal.

### 4. Creación del DataFrame

- Se crea un DataFrame de pandas llamado `df` con tres columnas:
  - `ID`: identificador único para cada persona (del 1 al 1000).
  - `Edad`: edad simulada.
  - `Estatura (cm)`: estatura simulada en centímetros.

### 5. Resumen estadístico

- Se imprime un resumen estadístico de las variables del DataFrame usando `df.describe()`, mostrando:
  - Cantidad de datos (`count`)
  - Media (`mean`)
  - Desviación estándar (`std`)
  - Valores mínimo y máximo (`min`, `max`)
  - Percentiles (25%, 50%, 75%)

### 6. Visualización: histogramas

- Se crean dos histogramas en una sola figura:
  - **Histograma de Edades**: muestra la frecuencia de las edades simuladas.
  - **Histograma de Estaturas**: muestra la frecuencia de las estaturas simuladas.
- Cada histograma tiene etiquetas y colores diferenciados para facilitar la interpretación visual.
- Finalmente, se ajusta el diseño y se muestra la figura.

### Resumen

- **Simula datos realistas** de edades y estaturas para 1000 personas.
- **Organiza los datos** en un DataFrame para análisis posterior.
- **Calcula y muestra estadísticas descriptivas** de las variables simuladas.
- **Visualiza la distribución** de los datos mediante histogramas, permitiendo observar la forma de campana característica de la distribución normal en ambas variables.

Este código es útil para practicar conceptos de estadística y análisis de datos, así como para ilustrar la generación y visualización de datos simulados.

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Configuración general
np.random.seed(42)
n = 1000

# Generación de datos simulados
edades = np.round(np.random.normal(21, 1.5, n)).astype(int)
estaturas = np.round(np.random.normal(175, 8, n), 1)

# Creación del DataFrame
df = pd.DataFrame({
    "ID": range(1, n + 1),
    "Edad": edades,
    "Estatura (cm)": estaturas
})

# Resumen estadístico
print(df.describe())

# Visualización: histogramas
fig, axs = plt.subplots(1, 2, figsize=(12, 5))

axs[0].hist(df["Edad"], bins=15, color='skyblue', edgecolor='black')
axs[0].set_title("Histograma de Edades")
axs[0].set_xlabel("Edad")
axs[0].set_ylabel("Frecuencia")

axs[1].hist(df["Estatura (cm)"], bins=20, color='salmon', edgecolor='black')
axs[1].set_title("Histograma de Estaturas")
axs[1].set_xlabel("Estatura (cm)")
axs[1].set_ylabel("Frecuencia")

plt.tight_layout()
plt.show()
````

## Cálculo de valores z


### Explicación del código

- **Parámetros de entrada:**  
  Se define una lista `edades_objetivo` con las edades para las que se desea calcular el valor z (en este caso, 25 y 17 años).

- **Cálculo de parámetros estadísticos:**  
  Se calcula la media (`media_edad`) y la desviación estándar (`desviacion_edad`) de la columna `Edad` del DataFrame `df`. Estos valores representan el promedio y la dispersión de las edades simuladas.

- **Función para el valor z:**  
  Se define la función `calcular_valor_z`, que aplica la fórmula estándar:
  $$
  z = \frac{X - \mu}{\sigma}
  $$
  donde $$ X $$ es la edad de interés, $$ \mu $$ es la media y $$ \sigma $$ es la desviación estándar.

- **Cálculo y presentación:**  
  El código recorre cada edad objetivo, calcula su valor z y lo imprime de forma clara y redondeada a dos decimales.




````python
# Calcular el valor z para edades específicas

# Parámetros de entrada
edades_objetivo = [25, 17]

# Cálculo de media y desviación estándar de la columna 'Edad'
media_edad = df['Edad'].mean()
desviacion_edad = df['Edad'].std()

# Función para calcular el valor z
def calcular_valor_z(edad, media, desviacion):
    return (edad - media) / desviacion

# Cálculo y presentación de resultados
for edad in edades_objetivo:
    z = calcular_valor_z(edad, media_edad, desviacion_edad)
    print(f"El valor z para una edad de {edad} años es: {z:.2f}")
````

## Límites de la regla empírica y visualización
### 1. ¿Qué hace este código?
El código calcula los intervalos en los que se encuentra aproximadamente el 68%, 95% y 99.7% de las estaturas de los encuestados, utilizando la **regla empírica** o **regla 68-95-99.7** de la distribución normal. Además, genera una gráfica que ilustra estos intervalos sobre el histograma de las estaturas.

### 2. Explicación pedagógica y funcionalidad
#### a) Fundamento teórico

- **Regla empírica:** En una distribución normal:
  - El 68% de los datos cae dentro de ±1 desviación estándar de la media.
  - El 95% dentro de ±2 desviaciones estándar.
  - El 99.7% dentro de ±3 desviaciones estándar.
- Esto permite estimar rápidamente la dispersión de los datos y detectar valores atípicos.

#### b) Pasos del código

1. **Cálculo de la media y desviación estándar**
   - Se obtiene la media y la desviación estándar de la columna `'Estatura (cm)'` del DataFrame.
2. **Cálculo de límites para cada intervalo**
   - Para cada porcentaje (68%, 95%, 99.7%), se calcula el límite inferior y superior aplicando la fórmula:
     $$
     \text{Límite inferior} = \text{media} - k \cdot \text{desviación estándar}
     $$
     $$
     \text{Límite superior} = \text{media} + k \cdot \text{desviación estándar}
     $$
     Donde $$ k = 1, 2, 3 $$ según el porcentaje.
3. **Impresión de resultados**
   - El código muestra los intervalos calculados para cada porcentaje.
4. **Visualización**
   - Se genera un histograma de las estaturas con una curva de densidad y líneas verticales que indican la media y los límites calculados para el 68%, 95% y 99.7%.

### 3. Visualización
El siguiente histograma muestra la distribución de estaturas, la media (línea roja) y los intervalos de la regla empírica con diferentes colores para facilitar su interpretación.

### 4. ¿Por qué es útil este análisis?
- Permite identificar rápidamente la dispersión y concentración de los datos.
- Facilita la detección de valores atípicos (outliers).
- Es una herramienta fundamental para interpretar resultados en estadística descriptiva y para la toma de decisiones basada en datos.

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Generar datos simulados
np.random.seed(42)
n = 1000
edades = np.round(np.random.normal(21, 1.5, n)).astype(int)
estaturas = np.round(np.random.normal(175, 8, n), 1)

df = pd.DataFrame({
    "ID": range(1, n + 1),
    "Edad": edades,
    "Estatura (cm)": estaturas
})

# 2. Calcular media y desviación estándar de las estaturas
media_estatura = df['Estatura (cm)'].mean()
desviacion_estatura = df['Estatura (cm)'].std()

# 3. Calcular los límites para los intervalos de la regla empírica
limites = {
    "68%": (media_estatura - desviacion_estatura, media_estatura + desviacion_estatura),
    "95%": (media_estatura - 2 * desviacion_estatura, media_estatura + 2 * desviacion_estatura),
    "99.7%": (media_estatura - 3 * desviacion_estatura, media_estatura + 3 * desviacion_estatura)
}

# 4. Mostrar los resultados en consola
for porcentaje, (lim_inf, lim_sup) in limites.items():
    print(f"El {porcentaje} de las estaturas se encuentra entre {lim_inf:.2f} cm y {lim_sup:.2f} cm")

# 5. Visualización: histograma con cortes de la regla empírica
plt.figure(figsize=(10, 6))
sns.histplot(df['Estatura (cm)'], kde=True, color='lightgray', edgecolor='black')

# Línea de la media
plt.axvline(media_estatura, color='red', linestyle='dashed', linewidth=2, label=f'Media: {media_estatura:.2f}')

# Líneas de los intervalos
colores = {'68%': 'green', '95%': 'blue', '99.7%': 'purple'}
for porcentaje, (lim_inf, lim_sup) in limites.items():
    plt.axvline(lim_inf, color=colores[porcentaje], linestyle='dashed', linewidth=2, label=f'{porcentaje} (inf): {lim_inf:.2f}')
    plt.axvline(lim_sup, color=colores[porcentaje], linestyle='dashed', linewidth=2, label=f'{porcentaje} (sup): {lim_sup:.2f}')

plt.title('Distribución de Estaturas con Cortes de la Regla Empírica')
plt.xlabel('Estatura (cm)')
plt.ylabel('Frecuencia')
plt.legend()
plt.tight_layout()
plt.show()
````

```markdown
## Determinación de áreas bajo la curva normal

En el análisis estadístico, la determinación de áreas bajo la curva normal es fundamental para calcular probabilidades asociadas a variables continuas que siguen una distribución normal. Este procedimiento permite responder preguntas como: ¿qué proporción de una población presenta valores menores o iguales a un umbral específico? O bien, ¿cuál es la probabilidad de que una variable aleatoria esté entre dos valores dados?

### Cálculo de probabilidades en la distribución normal

La probabilidad de que una variable aleatoria $$ X $$ tome un valor menor o igual a un cierto umbral $$ x $$ en una distribución normal se expresa como:

$$
P(X \leq x)
$$

Este valor corresponde al área bajo la curva normal desde menos infinito hasta $$ x $$.

#### Herramienta computacional: `norm.cdf` en Python

En Python, la función `norm.cdf` de la biblioteca `scipy.stats` permite calcular de manera eficiente esta probabilidad acumulada. Su sintaxis es la siguiente:

```
from scipy.stats import norm
probabilidad = norm.cdf(x, loc=media, scale=desviacion)
```

#### Parámetros de la función

- **`x`**: Valor de interés para el cual se desea calcular la probabilidad acumulada (por ejemplo, una estatura de 180 cm).
- **`loc`**: Media ($$ \mu $$) de la distribución normal.
- **`scale`**: Desviación estándar ($$ \sigma $$) de la distribución.

#### Ejemplo práctico

Suponga que la estatura de una población universitaria sigue una distribución normal con media de 175 cm y desviación estándar de 8 cm. Para determinar el porcentaje de personas con estatura menor o igual a 180 cm, se utiliza:

```
from scipy.stats import norm
prob = norm.cdf(180, loc=175, scale=8)
print(prob)  # Salida: 0.734 (73.4%)
```

Esto indica que aproximadamente el 73.4% de la población tiene una estatura igual o inferior a 180 cm.

### Aplicaciones comunes

- **Cálculo de probabilidades:** Determinar la proporción de observaciones por debajo de un valor específico.
- **Determinación de percentiles:** Utilizar la función inversa `norm.ppf` para encontrar el valor correspondiente a un determinado percentil.
- **Validación de supuestos estadísticos:** Verificar la normalidad de los datos en pruebas estadísticas.

### Resumen conceptual

La función `norm.cdf(x, µ, σ)` devuelve la probabilidad de que una variable aleatoria normal tome un valor menor o igual a $$ x $$, facilitando el análisis y la interpretación de fenómenos que siguen una distribución normal. Su uso es esencial en la estadística aplicada, tanto para la toma de decisiones como para la inferencia sobre poblaciones.
```


- **Ejemplo:**  
  Se convierte cada límite de edad (20 y 23 años) a su valor z, usando la fórmula:
  $$
  z = \frac{X - \mu}{\sigma}
  $$
  donde $$ X $$ es la edad de interés, $$ \mu $$ la media y $$ \sigma $$ la desviación estándar.

- **Uso de la normal estándar:**  
  La probabilidad de que una edad esté entre 20 y 23 años es la diferencia entre las probabilidades acumuladas hasta cada valor z:
  $$
  P(20 < X < 23) = P(Z < z_{\text{superior}}) - P(Z < z_{\text{inferior}})
  $$
  Esto se calcula con `norm.cdf(z_superior)` y `norm.cdf(z_inferior)`.

- **Ventaja de la estandarización:**  
  Permite generalizar el cálculo de probabilidades para cualquier intervalo, independientemente de la media y desviación estándar originales.

### Notas adicionales

- Si prefieres calcular la probabilidad directamente sobre la distribución original (sin convertir a z), puedes emplear:
  ```python
  probabilidad = norm.cdf(23, loc=media_edad, scale=desviacion_edad) - norm.cdf(20, loc=media_edad, scale=desviacion_edad)
  ```
  Ambos métodos son equivalentes si los parámetros están bien definidos.

- El resultado representa la fracción de la población simulada cuya edad está entre 20 y 23 años, bajo el supuesto de normalidad.

Este enfoque es fundamental en estadística para calcular áreas bajo la curva normal y resolver problemas de probabilidad con variables continuas.

````python
from scipy.stats import norm

# Supón que 'df' es tu DataFrame con la columna 'Edad'
# Cálculo de parámetros estadísticos
media_edad = df['Edad'].mean()
desviacion_edad = df['Edad'].std()

# Límites de edad
edad_inferior = 20
edad_superior = 23

# Estandarización (cálculo de valores z)
z_inferior = (edad_inferior - media_edad) / desviacion_edad
z_superior = (edad_superior - media_edad) / desviacion_edad

# Cálculo de la probabilidad usando la normal estándar
probabilidad = norm.cdf(z_superior) - norm.cdf(z_inferior)

print(f"La probabilidad de escoger una persona entre {edad_inferior} y {edad_superior} años es: {probabilidad:.4f}")

# sin estandarizar

probabilidadse = norm.cdf(23, loc=media_edad, scale=desviacion_edad) - norm.cdf(20, loc=media_edad, scale=desviacion_edad)

print(probabilidadse)
````

## Distribución Normal Inversa

La distribución normal inversa es una herramienta estadística esencial que permite determinar el valor específico de una variable aleatoria normal dado un nivel de probabilidad acumulada. Esto es especialmente útil para establecer umbrales o percentiles en una población, por ejemplo, para identificar la edad o estatura hasta la cual se encuentra un cierto porcentaje de individuos.

### Concepto fundamental

La función de distribución acumulada (CDF) de una variable aleatoria normal, denotada como \( P(X \leq x) \), calcula la probabilidad de que la variable tome un valor menor o igual a \( x \). La distribución normal inversa, o función inversa de la CDF, permite calcular el valor \( x \) correspondiente a una probabilidad acumulada dada \( q \). Matemáticamente, se busca:

\[
x = \text{norm.ppf}(q, \mu, \sigma)
\]

donde:

- \( q \) es la probabilidad acumulada (entre 0 y 1),
- \( \mu \) es la media de la distribución,
- \( \sigma \) es la desviación estándar.

### Funcionalidad de `norm.ppf`

- Es la inversa de la función de distribución acumulada (CDF).
- Calcula el valor \( x \) tal que la probabilidad acumulada hasta ese punto es igual a \( q \).
- Permite transformar una probabilidad en un valor de la variable aleatoria.

### Parámetros clave

| Parámetro | Descripción                          |
|-----------|------------------------------------|
| `q`       | Probabilidad acumulada (0 ≤ q ≤ 1) |
| `loc`     | Media de la distribución (\( \mu \)) |
| `scale`   | Desviación estándar (\( \sigma \))  |

### Ejemplo práctico

Supongamos que la estatura de una población sigue una distribución normal con media \( \mu = 175 \, \text{cm} \) y desviación estándar \( \sigma = 8 \, \text{cm} \). Para determinar la estatura que marca el percentil 90, es decir, el valor debajo del cual se encuentra el 90% de las observaciones, se utiliza:



````python
from scipy.stats import norm


# Cálculo de parámetros estadísticos
media_edad = df['Edad'].mean()
desviacion_edad = df['Edad'].std()

# Ejemplo 1: Edad correspondiente al percentil 4%
percentil_4 = 0.04
edad_percentil_4 = norm.ppf(percentil_4, loc=media_edad, scale=desviacion_edad)
print(f"El valor de la edad correspondiente al percentil 4% es: {edad_percentil_4:.2f} años")

# Ejemplo 2: Edad correspondiente al percentil 90%
percentil_90 = 0.90
edad_percentil_90 = norm.ppf(percentil_90, loc=media_edad, scale=desviacion_edad)
print(f"El valor de la edad correspondiente al percentil 90% es: {edad_percentil_90:.2f} años")
````

## Test de Normalidad

Determinar si un conjunto de datos sigue una distribución normal es fundamental en estadística, ya que muchos métodos y pruebas requieren este supuesto. A continuación se presentan los tests de normalidad más utilizados, sus características principales y ejemplos prácticos en Python.

---

### 1. Test de Shapiro-Wilk

- **Propósito:** Evalúa la hipótesis de que los datos provienen de una distribución normal, considerando simetría, curtosis y forma general.
- **Uso ideal:** Muestras pequeñas (n < 50).
- **Hipótesis:**
  - \( H_0 \): Los datos siguen una distribución normal.
  - \( H_1 \): Los datos no siguen una distribución normal.
- **Ventaja:** Alta potencia para muestras pequeñas.


---

### 2. Test de Kolmogorov-Smirnov (KS)

- **Propósito:** Compara la distribución empírica de los datos con una distribución normal teórica (ajustando media y desviación estándar).
- **Uso ideal:** Muestras grandes o cuando se conocen los parámetros de la distribución.
- **Limitación:** Menos sensible que Shapiro-Wilk para detectar desviaciones de la normalidad

### 3. Test de Anderson-Darling

- **Propósito:** Similar al KS, pero otorga mayor peso a las colas de la distribución.
- **Uso ideal:** Detectar desviaciones en las colas (útil en análisis de riesgos).
- **Ventaja:** Más sensible que KS en muestras moderadas.

### 4. Test de Lilliefors

- **Propósito:** Adaptación del KS para normalidad cuando la media y desviación estándar se estiman de los datos.
- **Uso ideal:** Muestras pequeñas a medianas.
- **Nota:** Es menos común en bibliotecas modernas, ya que Shapiro-Wilk suele preferirse.

---

### 5. Test de Jarque-Bera

- **Propósito:** Evalúa la normalidad considerando asimetría (simetría) y curtosis (peso de las colas).
- **Uso ideal:** Muestras grandes (n > 2000), común en econometría.
- **Fórmula:**
\[
JB = n \left( \frac{S^2}{6} + \frac{(K-3)^2}{24} \right)
\]
donde \( S \) es la asimetría y \( K \) la curtosis.

### Consejos prácticos

1. **Muestras pequeñas:** Prefiere Shapiro-Wilk o Lilliefors.
2. **Muestras grandes:** Usa Jarque-Bera y complementa con gráficos Q-Q.
3. **Detección de valores atípicos:** Anderson-Darling es más sensible en las colas.
4. **Siempre reporta el valor-p** y acompaña tus resultados con visualizaciones.

Estos tests y herramientas te permitirán evaluar de manera rigurosa la normalidad de tus datos antes de aplicar métodos estadísticos que dependen de este supuesto.

````python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
from scipy.stats import shapiro, kstest, anderson, jarque_bera, norm

np.random.seed(42)  # Reproducibilidad

# Generación de los conjuntos de datos
datos = {
    "Normal": np.random.normal(loc=0, scale=1, size=100),
    "Sesgado": np.random.lognormal(mean=0, sigma=0.8, size=100),
    "Colas Pesadas": np.random.standard_t(df=3, size=100),
    "Colas Ligeras": np.random.uniform(low=-3, high=3, size=100),
    "Outliers": np.concatenate([np.random.normal(0, 1, 90), np.random.normal(10, 1, 10)])
}

# Función para aplicar los tests de normalidad
def aplicar_tests(datos, nombre):
    print(f"\n===== {nombre} =====")
    stat_sw, p_sw = shapiro(datos)
    print(f"Shapiro-Wilk: Estadístico={stat_sw:.3f}, p={p_sw:.4f}")
    media, desv = np.mean(datos), np.std(datos)
    stat_ks, p_ks = kstest(datos, "norm", args=(media, desv))
    print(f"Kolmogorov-Smirnov: Estadístico={stat_ks:.3f}, p={p_ks:.4f}")
    resultado_ad = anderson(datos)
    print(f"Anderson-Darling: Estadístico={resultado_ad.statistic:.3f}, Valor crítico (5%): {resultado_ad.critical_values[2]:.3f}")
    stat_jb, p_jb = jarque_bera(datos)
    print(f"Jarque-Bera: Estadístico={stat_jb:.3f}, p={p_jb:.4f}")

# Aplicar los tests y mostrar conclusiones pedagógicas
conclusiones = {}

for nombre, datos_i in datos.items():
    aplicar_tests(datos_i, nombre)
    # Conclusión pedagógica por tipo de datos
    if nombre == "Normal":
        conclusiones[nombre] = (
            "Los datos normales presentan valores p altos en los tests, lo que indica que no se rechaza la hipótesis de normalidad. "
            "En los gráficos, el histograma y la curva de densidad se asemejan a la campana de Gauss, y el gráfico Q-Q muestra los puntos alineados en la diagonal."
        )
    elif nombre == "Sesgado":
        conclusiones[nombre] = (
            "Los datos sesgados muestran valores p bajos, rechazando la normalidad. El histograma es asimétrico (sesgo positivo), "
            "la curva de densidad se aleja de la normal teórica y el gráfico Q-Q se curva, evidenciando la falta de normalidad."
        )
    elif nombre == "Colas Pesadas":
        conclusiones[nombre] = (
            "Las colas pesadas generan valores p bajos y estadísticos altos, indicando no normalidad. El histograma muestra colas largas, "
            "y el gráfico Q-Q exhibe desviaciones en los extremos (colas), típico de distribuciones t con pocos grados de libertad."
        )
    elif nombre == "Colas Ligeras":
        conclusiones[nombre] = (
            "Los datos uniformes tienen colas ligeras y valores p bajos en los tests, rechazando la normalidad. El histograma es plano, "
            "no hay forma de campana y el gráfico Q-Q se aparta de la diagonal en los extremos."
        )
    elif nombre == "Outliers":
        conclusiones[nombre] = (
            "La presencia de outliers reduce drásticamente los valores p y aumenta los estadísticos de los tests, señalando fuerte no normalidad. "
            "El histograma muestra dos grupos y el gráfico Q-Q revela puntos alejados de la diagonal en los extremos."
        )

# Visualización: histogramas y gráficos Q-Q
fig, axes = plt.subplots(nrows=5, ncols=2, figsize=(12, 20))
for i, (nombre, datos_i) in enumerate(datos.items()):
    # Histograma + curva de densidad y normal teórica
    sns.histplot(datos_i, kde=True, stat="density", ax=axes[i, 0], color="skyblue", label="Datos")
    x = np.linspace(min(datos_i), max(datos_i), 100)
    media, desv = np.mean(datos_i), np.std(datos_i)
    y = norm.pdf(x, loc=media, scale=desv)
    axes[i, 0].plot(x, y, color="red", linestyle="--", label="Normal Teórica")
    axes[i, 0].set_title(f"{nombre} - Histograma y Densidad")
    axes[i, 0].legend()
    # Gráfico Q-Q
    sm.qqplot(datos_i, line="s", ax=axes[i, 1], color="blue")
    axes[i, 1].set_title(f"{nombre} - Gráfico Q-Q")
plt.tight_layout()
plt.show()
````

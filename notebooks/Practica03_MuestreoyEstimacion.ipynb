## Inferencia Estadística: Concepto y Razones para el Muestreo

La **inferencia estadística** consiste en identificar patrones o características de una población a partir del análisis de una muestra representativa de dicha población. Este enfoque se utiliza principalmente por las siguientes razones:

1. Contactar a toda la población requiere demasiado tiempo.
2. El costo de analizar a toda la población es muy elevado.
3. En muchos casos, es imposible verificar a todos los individuos de la población.
4. Algunas pruebas necesarias pueden ser de naturaleza destructiva y no se pueden aplicar a todos los elementos.
5. Los resultados obtenidos a partir de la muestra suelen ser suficientemente precisos y representativos.

---

## Tipos de Muestreo

- **Muestreo aleatorio simple:** Se selecciona una muestra de manera que cada elemento o individuo de la población tiene la misma probabilidad de ser escogido.

- **Muestreo aleatorio sistemático:** Se elige un punto de inicio aleatorio y luego se selecciona cada k-ésimo elemento de la población. Por ejemplo, para una muestra de 100 personas, se determina un valor aleatorio inicial y se selecciona cada k-ésimo individuo según la población.

- **Muestreo aleatorio estratificado:** La población se divide en subgrupos llamados estratos, y se selecciona una muestra aleatoria de cada estrato para asegurar representación proporcional.

- **Muestreo por conglomerados:** Se utiliza cuando la población está distribuida geográficamente. Se seleccionan grupos o conglomerados completos según la distribución regional para formar la muestra.

- **Error de muestreo:** Es la diferencia entre el estadístico calculado a partir de la muestra y el parámetro real de la población, reflejando la variabilidad inherente al proceso de muestreo.


```python
# Ejemplo de muestreo simple 

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Configuración para reproducibilidad
np.random.seed(42)

# Crear DataFrame con 1000 puntajes (media=75, desviación estándar=10)
puntajes = np.random.normal(loc=75, scale=10, size=1000)
df = pd.DataFrame({'Puntaje': puntajes})

# Generar muestra aleatoria de 150 puntajes
muestra = df.sample(n=150, random_state=42)

# Estadísticas descriptivas
print("Estadísticas descriptivas de la población:")
print(df.describe())

print("\nEstadísticas descriptivas de la muestra:")
print(muestra.describe())

# Histograma comparativo
plt.figure(figsize=(12, 6))
plt.hist(df['Puntaje'], bins=30, alpha=0.5, label='Población (1000)')
plt.hist(muestra['Puntaje'], bins=30, alpha=0.7, label='Muestra (150)')
plt.title('Histograma de Puntajes: Población vs Muestra')
plt.xlabel('Puntaje')
plt.ylabel('Frecuencia')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Muestreo aleatorio sistemático

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Configuración para reproducibilidad
np.random.seed(42)

# Crear DataFrame con 1000 puntajes (media=75, desviación estándar=10)
puntajes = np.random.normal(loc=75, scale=10, size=1000)
df = pd.DataFrame({'Puntaje': puntajes})

# Función de muestreo aleatorio sistemático
def systematic_sampling(df, sample_size):
    interval = len(df) // sample_size
    start = np.random.randint(0, interval)
    indices = np.arange(start, len(df), interval)
    return df.iloc[indices]

# Obtener muestra sistemática de 100 puntajes
muestra_sistematica = systematic_sampling(df, 100)

# Estadísticas descriptivas
print("Estadísticas descriptivas de la población:")
print(df['Puntaje'].describe())

print("\nEstadísticas descriptivas de la muestra sistemática:")
print(muestra_sistematica['Puntaje'].describe())

# Histograma comparativo
plt.figure(figsize=(12, 6))
plt.hist(df['Puntaje'], bins=30, alpha=0.5, label='Población (1000)')
plt.hist(muestra_sistematica['Puntaje'], bins=30, alpha=0.7, label='Muestra Sistemática (100)')
plt.title('Histograma de Puntajes: Población vs Muestra Sistemática')
plt.xlabel('Puntaje')
plt.ylabel('Frecuencia')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Configuración para reproducibilidad
np.random.seed(42)

# Crear DataFrame con 1000 puntajes (media=75, desviación estándar=10)
puntajes = np.random.normal(loc=75, scale=10, size=1000)
df = pd.DataFrame({'Puntaje': puntajes})

# Definir los estratos
bins = [0, 60, 75, 90, 100]
labels = ['Deficiente', 'Básico', 'Sobresaliente', 'Excelente']
df['Estrato'] = pd.cut(df['Puntaje'], bins=bins, labels=labels, right=False)

# Verificar cantidad de datos por estrato
print(df['Estrato'].value_counts().sort_index())

# Tomar muestra de 25 puntajes por estrato (si hay suficientes datos)
muestra_estratificada = df.groupby('Estrato').apply(lambda x: x.sample(n=25, random_state=42) if len(x) >= 25 else x)
muestra_estratificada = muestra_estratificada.reset_index(drop=True)

# Estadísticas descriptivas por estrato
print(muestra_estratificada.groupby('Estrato')['Puntaje'].describe())

# Histograma comparativo por estrato
plt.figure(figsize=(12, 6))
for label in labels:
    subset_poblacion = df[df['Estrato'] == label]['Puntaje']
    subset_muestra = muestra_estratificada[muestra_estratificada['Estrato'] == label]['Puntaje']
    plt.hist(subset_poblacion, bins=15, alpha=0.3, label=f'Población {label}')
    plt.hist(subset_muestra, bins=15, alpha=0.7, label=f'Muestra {label}')
plt.title('Histograma de Puntajes por Estrato: Población vs Muestra Estratificada')
plt.xlabel('Puntaje')
plt.ylabel('Frecuencia')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np

# Configuración para reproducibilidad
np.random.seed(42)

# --- Población de puntajes ---
poblacion_puntajes = np.random.normal(loc=75, scale=10, size=1000)
poblacion_media = np.mean(poblacion_puntajes)
poblacion_desviacion = np.std(poblacion_puntajes)

# --- Muestreo aleatorio simple ---
muestra_aleatoria = pd.DataFrame({'Puntaje': poblacion_puntajes}).sample(n=150, random_state=42)
error_aleatorio = abs(muestra_aleatoria['Puntaje'].mean() - poblacion_media)
print("Muestreo Aleatorio Simple:")
print(f"  Media de la muestra: {muestra_aleatoria['Puntaje'].mean():.2f}")
print(f"  Desviación estándar de la muestra: {muestra_aleatoria['Puntaje'].std():.2f}")
print(f"  Media de la población: {poblacion_media:.2f}")
print(f"  Desviación estándar de la población: {poblacion_desviacion:.2f}")
print(f"  Error de muestreo: {error_aleatorio:.2f}")

# --- Muestreo aleatorio sistemático ---
def systematic_sampling(df, sample_size):
    interval = len(df) // sample_size
    start = np.random.randint(0, interval)
    indices = np.arange(start, len(df), interval)
    return df.iloc[indices]

df_puntajes = pd.DataFrame({'Puntaje': poblacion_puntajes})
muestra_sistematica = systematic_sampling(df_puntajes, 100)
error_sistematico = abs(muestra_sistematica['Puntaje'].mean() - poblacion_media)
print("\nMuestreo Sistemático:")
print(f"  Media de la muestra: {muestra_sistematica['Puntaje'].mean():.2f}")
print(f"  Desviación estándar de la muestra: {muestra_sistematica['Puntaje'].std():.2f}")
print(f"  Media de la población: {poblacion_media:.2f}")
print(f"  Desviación estándar de la población: {poblacion_desviacion:.2f}")
print(f"  Error de muestreo: {error_sistematico:.2f}")

# --- Muestreo aleatorio estratificado ---
bins = [0, 60, 75, 90, 100]
labels = ['Deficiente', 'Básico', 'Sobresaliente', 'Excelente']
df_puntajes['Estrato'] = pd.cut(df_puntajes['Puntaje'], bins=bins, labels=labels, right=False)
muestra_estratificada = df_puntajes.groupby('Estrato').apply(lambda x: x.sample(n=25, random_state=42) if len(x) >= 25 else x)
muestra_estratificada = muestra_estratificada.reset_index(drop=True)
error_estratificado = abs(muestra_estratificada['Puntaje'].mean() - poblacion_media)
print("\nMuestreo Estratificado:")
print(f"  Media de la muestra: {muestra_estratificada['Puntaje'].mean():.2f}")
print(f"  Desviación estándar de la muestra: {muestra_estratificada['Puntaje'].std():.2f}")
print(f"  Media de la población: {poblacion_media:.2f}")
print(f"  Desviación estándar de la población: {poblacion_desviacion:.2f}")
print(f"  Error de muestreo: {error_estratificado:.2f}")

# --- Muestreo por conglomerados (alturas por país) ---
# Datos realistas de altura promedio (cm) y desviación estándar por país
paises = {
    'Estados Unidos': {'media': 175, 'desviacion': 7.5},
    'Japón': {'media': 170, 'desviacion': 6.5},
    'Países Bajos': {'media': 183, 'desviacion': 7.0},
    'México': {'media': 167, 'desviacion': 7.0},
    'India': {'media': 165, 'desviacion': 6.0}
}
total_muestras = 100
muestras_por_pais = total_muestras // len(paises)
np.random.seed(42)
data = []
for pais, stats in paises.items():
    alturas = np.random.normal(loc=stats['media'], scale=stats['desviacion'], size=muestras_por_pais)
    for altura in alturas:
        data.append({'País': pais, 'Altura': altura})
df_alturas = pd.DataFrame(data)
# Parámetros poblacionales (toda la muestra generada)
media_poblacional_altura = df_alturas['Altura'].mean()
desv_poblacional_altura = df_alturas['Altura'].std()
# Ejemplo: seleccionamos un país como conglomerado (México)
muestra_conglomerado = df_alturas[df_alturas['País'] == 'México']
error_conglomerado = abs(muestra_conglomerado['Altura'].mean() - media_poblacional_altura)
print("\nMuestreo por Conglomerados (México):")
print(f"  Media de la muestra: {muestra_conglomerado['Altura'].mean():.2f}")
print(f"  Desviación estándar de la muestra: {muestra_conglomerado['Altura'].std():.2f}")
print(f"  Media de la población: {media_poblacional_altura:.2f}")
print(f"  Desviación estándar de la población: {desv_poblacional_altura:.2f}")
print(f"  Error de muestreo: {error_conglomerado:.2f}")
```

3. Teorema del límite central

# Explicación del Teorema del Límite Central (TLC)

El **Teorema del Límite Central (TLC)** es fundamental para entender por qué muchas técnicas estadísticas funcionan, incluso cuando los datos originales no son "perfectos". Aquí tienes una explicación clara y revisada, paso a paso.

---

## 1. ¿Qué dice el TLC?

Imagina que tomas **muchas muestras** (por ejemplo, 1000) del mismo tamaño ($n$) de **cualquier población** (no importa si es sesgada, uniforme, bimodal, etc.). Para cada muestra, calculas su **media ($\bar{x}$)**. Según el TLC:

- **La distribución de todas esas medias muestrales tiende a ser normal** (curva de campana), incluso si la población original no lo es.
- **Cuanto mayor sea el tamaño de la muestra ($n$), mejor será la aproximación a la normalidad**.

---

## 2. Ejemplo Visual

Supón que analizas el **ingreso mensual** en una ciudad, que suele tener una distribución **sesgada a la derecha** (muchas personas con ingresos bajos y pocas con ingresos altos).

- **Paso 1:** Toma 1000 muestras de tamaño $n = 5$, calcula sus medias y gráficas. La distribución será **poco normal**, quizás aún sesgada.
- **Paso 2:** Repite el proceso con muestras de $n = 30$. Ahora, la distribución de las medias será **más simétrica y acampanada**.
- **Paso 3:** Si usas $n = 50$, la curva será casi **perfectamente normal**.

**Conclusión:** ¡Las medias tienden a la normalidad aunque la población original no lo sea!

---

## 3. ¿Por qué ocurre esto?

- **Efecto promedio:** Al calcular la media, los valores extremos se diluyen. En una muestra grande, es poco probable que todos los datos sean muy altos o muy bajos.
- **Ley de los Grandes Números:** A mayor $n$, la media muestral ($\bar{x}$) se acerca más a la media poblacional ($\mu$).
- **Fórmulas clave:**
  - Media del muestreo: $\mu_{\bar{x}} = \mu$ (igual a la media poblacional).
  - Error estándar: $\sigma_{\bar{x}} = \frac{\sigma}{\sqrt{n}}$ (la dispersión disminuye al aumentar $n$).

---

## 4. Condiciones Importantes

- **Muestras aleatorias e independientes.**
- **Tamaño de muestra suficientemente grande:**
  - Para poblaciones no normales, se recomienda $n \geq 30$ (regla práctica).
  - Si la población es casi normal, con $n = 15$ podría bastar.

---

## 5. ¿Para qué sirve el TLC?

- **Inferencia estadística:** Permite usar métodos que asumen normalidad (pruebas $t$, ANOVA, intervalos de confianza) incluso con poblaciones no normales.
- **Predicciones:** Podemos calcular probabilidades sobre medias muestrales usando la distribución normal.

---

## 6. Errores Comunes

- **Confundir la distribución de la población con la de las medias:** El TLC se refiere a las medias, no a los datos originales.
- **Creer que $n = 30$ siempre es suficiente:** Si la población es muy extraña (por ejemplo, fuertemente bimodal), puede requerirse un $n$ mayor.

---

## 7. Analogía Cotidiana

Piensa en lanzar un dado:
- **Un lanzamiento:** Distribución uniforme (probabilidades iguales para 1-6).
- **Promedio de 50 lanzamientos:** Si repites esto muchas veces, los promedios se distribuirán normalmente alrededor de 3.5.

---

## Resumen Final

El TLC es como un "superpoder" estadístico:
1. **Transforma cualquier distribución en normal** para las medias muestrales.
2. **Funciona mejor con muestras grandes.**
3. **Es la base para hacer predicciones y conclusiones** sobre poblaciones usando muestras.

¡Ahora puedes entender por qué tantos métodos estadísticos son válidos incluso cuando los datos reales no son perfectos!


```python
# Teorema del Límite Central: Simulación y Visualización con Población Bimodal

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import norm

# 1. Configuración inicial
np.random.seed(42)  # Para reproducibilidad
sns.set_style('darkgrid')  # Estilo de Seaborn para gráficos

# 2. Crear una población bimodal no normal
poblacion = np.concatenate((
    np.random.normal(loc=10, scale=2, size=50000),  # Primera moda
    np.random.normal(loc=30, scale=5, size=50000)   # Segunda moda
))

# 3. Parámetros de la población
mu_poblacion = np.mean(poblacion)
sigma_poblacion = np.std(poblacion, ddof=1)  # ddof=1 para la desviación estándar muestral

# 4. Función para simular el TLC
def simular_tlc(poblacion, tamano_muestras, n_experimentos=1000):
    """
    Simula el Teorema del Límite Central extrayendo medias muestrales de la población.
    Usa muestreo con reemplazo para evitar errores cuando n es grande.
    """
    medias = []
    for _ in range(n_experimentos):
        muestra = np.random.choice(poblacion, size=tamano_muestras, replace=True)
        medias.append(np.mean(muestra))
    return medias

# 5. Tamaños de muestra a comparar
tamanos_muestra = [1, 5, 30, 50]

# 6. Crear gráficos
fig, axs = plt.subplots(2, 2, figsize=(12, 10))
axs = axs.flatten()

for i, n in enumerate(tamanos_muestra):
    medias = simular_tlc(poblacion, n)

    # Histograma de medias muestrales con KDE
    sns.histplot(medias, kde=True, ax=axs[i], stat='density', color='skyblue', edgecolor='black')

    # Curva normal teórica
    xmin, xmax = axs[i].get_xlim()
    x = np.linspace(xmin, xmax, 100)
    mu_teorico = mu_poblacion
    sigma_teorico = sigma_poblacion / np.sqrt(n)
    y = norm.pdf(x, mu_teorico, sigma_teorico)

    axs[i].plot(x, y, 'r--', linewidth=2, label='Normal teórica')
    axs[i].set_title(f'n = {n}\n$\mu={mu_teorico:.2f}$, $\sigma={sigma_teorico:.2f}$')
    axs[i].set_xlabel('Media muestral')
    axs[i].set_ylabel('Densidad')
    axs[i].legend()

plt.suptitle('Teorema del Límite Central - Distribución de Medias Muestrales', fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.96])
plt.show()
```

5. Estimadores puntuales e intervalos de confianza

# Estimadores Puntuales e Intervalos de Confianza

La **estimación estadística** es fundamental para inferir características desconocidas de una población a partir de una muestra. Existen dos conceptos clave: los **estimadores puntuales** y los **intervalos de confianza**.

## 1. Estimador Puntual

Un **estimador puntual** es un valor único calculado a partir de los datos de una muestra, que se utiliza para aproximar un parámetro desconocido de la población. Los estimadores puntuales más comunes son:

- **Media muestral ($\overline{X}$):** Estima la media poblacional ($\mu$).
- **Proporción muestral ($\hat{p}$):** Estima la proporción poblacional ($p$).
- **Desviación estándar muestral ($s$):** Estima la desviación estándar poblacional ($\sigma$).

> **Ejemplo:**  
> El turismo es una fuente importante de ingresos para Barbados. La Oficina de Turismo desea estimar el gasto promedio de los turistas. Como no es posible encuestar a todos, selecciona aleatoriamente a 500 turistas al salir del país y les pregunta cuánto gastaron.  
>  
> La **media de los gastos de estos 500 turistas** ($\overline{X}$) es un **estimador puntual** de la media poblacional ($\mu$), es decir, del gasto promedio real de todos los turistas.

## 2. Intervalos de Confianza

Un **intervalo de confianza** proporciona un rango de valores dentro del cual es probable que se encuentre el parámetro poblacional, con un cierto nivel de confianza (por ejemplo, 95%).

- **Interpretación:** Un intervalo de confianza del 95% significa que, si repitiéramos el muestreo muchas veces, aproximadamente el 95% de los intervalos calculados contendrían el valor real del parámetro poblacional.

### 2.1. Casos según la desviación estándar

Dependiendo de si se conoce o no la desviación estándar de la población ($\sigma$), se utilizan diferentes fórmulas:

| Caso | Fórmula del Intervalo de Confianza para la Media |
|------|-------------------------------------------------|
| **Se conoce $\sigma$** | $\overline{X} \pm z^* \left( \dfrac{\sigma}{\sqrt{n}} \right)$ |
| **No se conoce $\sigma$** | $\overline{X} \pm t^* \left( \dfrac{s}{\sqrt{n}} \right)$ |

- $z^*$: valor crítico de la normal estándar (usualmente 1.96 para 95% de confianza).
- $t^*$: valor crítico de la distribución t de Student (depende de $n-1$ grados de libertad).
- $n$: tamaño de la muestra.
- $s$: desviación estándar muestral.

### 2.2. Ejemplo aplicado

Siguiendo el ejemplo anterior:

- **Estimador puntual:** $\overline{X}$ = gasto medio de la muestra de 500 turistas.
- **Intervalo de confianza:**  
  Si la desviación estándar poblacional es desconocida (lo más común), se calcula:

  $$
  \overline{X} \pm t^* \left( \frac{s}{\sqrt{500}} \right)
  $$

  Esto da un rango donde es probable que esté el gasto promedio real de todos los turistas.

## 3. Resumen

- El **estimador puntual** es un valor único que estima un parámetro poblacional.
- El **intervalo de confianza** da un rango plausible para el parámetro, con un nivel de confianza especificado.
- Es crucial distinguir entre conocer o no la desviación estándar de la población para elegir la fórmula adecuada.
- Ambos conceptos permiten tomar decisiones informadas y cuantificar la incertidumbre inherente al muestreo.

**En conclusión:**  
Los estimadores puntuales y los intervalos de confianza son herramientas esenciales para inferir parámetros poblacionales y comunicar la precisión de nuestras estimaciones a partir de datos muestrales.


```python
# Solución del ejemplo anterior 

import numpy as np
import scipy.stats as stats

# Datos proporcionados
media_muestral = 78
desviacion_estandar = 12
tamano_muestra = 100

# Cálculo del error estándar de la media
error_estandar = desviacion_estandar / np.sqrt(tamano_muestra)

# Intervalos de confianza para los niveles más comunes
intervalo_confianza_95 = stats.norm.interval(0.95, loc=media_muestral, scale=error_estandar)
intervalo_confianza_90 = stats.norm.interval(0.90, loc=media_muestral, scale=error_estandar)
intervalo_confianza_99 = stats.norm.interval(0.99, loc=media_muestral, scale=error_estandar)

print(f"Intervalo de confianza del 95%: {intervalo_confianza_95}")
print(f"Intervalo de confianza del 90%: {intervalo_confianza_90}")
print(f"Intervalo de confianza del 99%: {intervalo_confianza_99}")

# Valores críticos de la normal estándar para los niveles más usados
critical_value_95 = stats.norm.ppf(0.975)  # 0.975 para intervalo bilateral al 95%
critical_value_90 = stats.norm.ppf(0.95)   # 0.95 para intervalo bilateral al 90%
critical_value_99 = stats.norm.ppf(0.995)  # 0.995 para intervalo bilateral al 99%

print("Valor crítico 95%:", critical_value_95)
print("Valor crítico 90%:", critical_value_90)
print("Valor crítico 99%:", critical_value_99)
```

# Ejemplos de Intervalos de Confianza para la Media Poblacional (\(\mu\)) con \(\sigma\) Conocida

Cuando la **desviación estándar poblacional** (\(\sigma\)) es conocida, se utiliza la **distribución normal (Z)** para calcular intervalos de confianza para la media poblacional. A continuación se presentan tres ejemplos detallados y una tabla comparativa.

---

## Fórmula General del Intervalo de Confianza

\[
\text{IC} = \bar{x} \pm Z_{\alpha/2} \cdot \frac{\sigma}{\sqrt{n}}
\]

Donde:
- \(\bar{x}\): Media muestral
- \(\sigma\): Desviación estándar poblacional (conocida)
- \(n\): Tamaño de la muestra
- \(Z_{\alpha/2}\): Valor crítico de la normal estándar para el nivel de confianza deseado

---

## Ejemplo 1: Control de Calidad en una Fábrica

**Contexto:**  
Una fábrica produce tornillos cuya longitud media es desconocida, pero se sabe que la desviación estándar poblacional es \(\sigma = 0.2\) cm. Se toma una muestra de 50 tornillos y se obtiene una media de \(\bar{x} = 5.6\) cm.

- **Nivel de confianza:** 95% (\(\alpha = 0.05\))
- **Valor crítico:** \(Z_{0.025} = 1.96\)
- **Error estándar:** \(\frac{0.2}{\sqrt{50}} = 0.0283\)
- **Margen de error:** \(1.96 \times 0.0283 = 0.0555\)
- **Intervalo de confianza:** \(5.6 \pm 0.0555 \to [5.5445,\ 5.6555]\)

**Interpretación:**  
Con un 95% de confianza, la longitud media de todos los tornillos producidos está entre **5.54 cm** y **5.66 cm**.

---

## Ejemplo 2: Rendimiento Académico

**Contexto:**  
Un colegio desea estimar la calificación promedio en matemáticas. La desviación estándar histórica es \(\sigma = 12\) puntos. En una muestra de 100 estudiantes, la media fue \(\bar{x} = 78\) puntos.

- **Nivel de confianza:** 90% (\(\alpha = 0.10\))
- **Valor crítico:** \(Z_{0.05} = 1.645\)
- **Error estándar:** \(\frac{12}{\sqrt{100}} = 1.2\)
- **Margen de error:** \(1.645 \times 1.2 = 1.974\)
- **Intervalo de confianza:** \(78 \pm 1.974 \to [76.026,\ 79.974]\)

**Interpretación:**  
Con un 90% de confianza, la calificación promedio en matemáticas está entre **76.03** y **79.97** puntos.

---

## Ejemplo 3: Mediciones Médicas

**Contexto:**  
Un hospital estima la presión arterial sistólica media de pacientes con diabetes. Se sabe que \(\sigma = 8\) mmHg. En una muestra de 30 pacientes, la media fue \(\bar{x} = 130\) mmHg.

- **Nivel de confianza:** 99% (\(\alpha = 0.01\))
- **Valor crítico:** \(Z_{0.005} = 2.576\)
- **Error estándar:** \(\frac{8}{\sqrt{30}} = 1.461\)
- **Margen de error:** \(2.576 \times 1.461 = 3.764\)
- **Intervalo de confianza:** \(130 \pm 3.764 \to [126.236,\ 133.764]\)

**Interpretación:**  
Con un 99% de confianza, la presión arterial sistólica media está entre **126.24 mmHg** y **133.76 mmHg**.

## Conclusiones Clave

- **A mayor nivel de confianza**, el intervalo es más amplio (Ejemplo 3 vs Ejemplo 2).
- **A mayor tamaño de muestra (\(n\))**, el margen de error disminuye (Ejemplo 2 vs Ejemplo 1).
- Se utiliza la **distribución Z** porque \(\sigma\) es conocida. Si fuera desconocida, se emplearía la distribución \(t\).



```python
import numpy as np
import scipy.stats as stats

def intervalo_confianza_z(media_muestral, sigma, n, confianza):
    """
    Calcula el intervalo de confianza para la media poblacional
    usando la distribución normal (Z), cuando sigma es conocida.

    Parámetros:
    - media_muestral: media de la muestra
    - sigma: desviación estándar poblacional (conocida)
    - n: tamaño de la muestra
    - confianza: nivel de confianza (ej. 0.95 para 95%)

    Retorna:
    - (limite_inferior, limite_superior): tupla con los límites del IC
    - z_critico: valor crítico Z
    - margen_error: margen de error calculado
    """
    alpha = 1 - confianza
    z_critico = stats.norm.ppf(1 - alpha/2)
    error_estandar = sigma / np.sqrt(n)
    margen_error = z_critico * error_estandar
    limite_inferior = media_muestral - margen_error
    limite_superior = media_muestral + margen_error
    return (limite_inferior, limite_superior), z_critico, margen_error

# Datos de los ejemplos
ejemplos = [
    {'nombre': 'Fábrica', 'media': 5.6, 'sigma': 0.2, 'n': 50, 'confianza': 0.95},
    {'nombre': 'Colegio', 'media': 78, 'sigma': 12, 'n': 100, 'confianza': 0.90},
    {'nombre': 'Hospital', 'media': 130, 'sigma': 8, 'n': 30, 'confianza': 0.99}
]

# Calcular y mostrar resultados
for ejemplo in ejemplos:
    ic, z, me = intervalo_confianza_z(
        ejemplo['media'], ejemplo['sigma'], ejemplo['n'], ejemplo['confianza']
    )
    print(f"Ejemplo: {ejemplo['nombre']}")
    print(f"  Media muestral: {ejemplo['media']}")
    print(f"  Sigma (desviación estándar poblacional): {ejemplo['sigma']}")
    print(f"  Tamaño de muestra: {ejemplo['n']}")
    print(f"  Nivel de confianza: {int(ejemplo['confianza']*100)}%")
    print(f"  Valor crítico Z: {z:.3f}")
    print(f"  Margen de error: {me:.3f}")
    print(f"  Intervalo de confianza: [{ic[0]:.3f}, {ic[1]:.3f}]")
    print("-" * 50)
```


## Ejemplo paso a paso: Intervalo de confianza para la media poblacional (\(\mu\)) con \(\sigma\) conocida

---

### **Contexto del ejemplo**
Una empresa de alimentos desea estimar el **peso promedio** de las bolsas de arroz que produce.

- **Desviación estándar poblacional (\(\sigma\))**: 15 gramos (dato histórico)
- **Tamaño de la muestra (\(n\))**: 20 bolsas de arroz
- **Nivel de confianza**: 95%

---

### **Paso 1: Datos de la muestra**
Pesos (en gramos) de las 20 bolsas:
[498, 502, 495, 510, 505, 500, 497, 503, 498, 506,
 501, 499, 504, 496, 507, 502, 493, 508, 495, 509]
---

### **Paso 2: Calcular la media muestral (\(\bar{x}\))**
Sumamos los pesos y dividimos entre el tamaño de la muestra (\(n = 20\)):
\[
\bar{x} = \frac{498 + 502 + \ldots + 509}{20} = \frac{10,\!030}{20} = 501.5 \text{ gramos}
\]

---

### **Paso 3: Fórmula del intervalo de confianza**
\[
\text{IC} = \bar{x} \pm Z_{\alpha/2} \cdot \frac{\sigma}{\sqrt{n}}
\]

---

### **Paso 4: Calcular el error estándar y el margen de error**

- **Error estándar**:
  \[
  \frac{\sigma}{\sqrt{n}} = \frac{15}{\sqrt{20}} \approx 3.354 \text{ gramos}
  \]

- **Valor crítico \(Z_{\alpha/2}\) para 95% de confianza**:
  \[
  Z_{\alpha/2} = 1.96
  \]

- **Margen de error (ME)**:
  \[
  ME = 1.96 \times 3.354 \approx 6.57 \text{ gramos}
  \]

---

### **Paso 5: Construir el intervalo de confianza**
\[
\text{IC} = 501.5 \pm 6.57 \implies [494.93,\ 508.07] \text{ gramos}
\]

---

### **Interpretación**
Con un **95% de confianza**, el peso promedio verdadero de todas las bolsas de arroz producidas está entre **494.93 gramos** y **508.07 gramos**.

---

### **Validación y notas finales**

- **¿Por qué usar la distribución Z?**
  - Porque la desviación estándar poblacional (\(\sigma\)) es conocida y la población se asume normal (o la muestra es suficientemente grande; aquí \(n=20\) es aceptable si la población no es muy sesgada).

- **¿Qué hacer si \(\sigma\) es desconocida?**
  - Se utiliza la distribución \(t\) de Student con \(n-1\) grados de libertad.

- **Aplicación práctica:**
  - La empresa puede estar razonablemente segura de que el peso promedio no está por debajo de 494.93 gramos, lo que ayuda a cumplir con regulaciones y estándares de calidad.

---


```python
# Ejemplo: Intervalo de confianza para la media poblacional (μ) con σ conocida

import math

# Paso 1: Datos de la muestra
pesos = [
    498, 502, 495, 510, 505, 500, 497, 503, 498, 506,
    501, 499, 504, 496, 507, 502, 493, 508, 495, 509
]
sigma = 15  # desviación estándar poblacional
n = len(pesos)  # tamaño de la muestra
nivel_confianza = 0.95

# Paso 2: Calcular la media muestral
media_muestral = sum(pesos) / n

# Paso 3 y 4: Calcular el error estándar y el margen de error
error_estandar = sigma / math.sqrt(n)
z_alpha_2 = 1.96  # Valor crítico para 95% de confianza
margen_error = z_alpha_2 * error_estandar

# Paso 5: Construir el intervalo de confianza
limite_inferior = media_muestral - margen_error
limite_superior = media_muestral + margen_error

# Mostrar resultados
print(f"Media muestral: {media_muestral:.1f} gramos")
print(f"Error estándar: {error_estandar:.3f} gramos")
print(f"Margen de error: {margen_error:.2f} gramos")
print(f"Intervalo de confianza al 95%: [{limite_inferior:.2f}, {limite_superior:.2f}] gramos")
```

En su clase de 40 alumnos de las 10 de la mañana, el número medio de palabras con faltas de ortografía fue de 6.05. Construya un intervalo de confianza de 95% para el número medio de palabras con faltas de ortografía en la población de ensayos.

## Desviación estándar poblacional $\sigma$ desconocida

En muchas situaciones prácticas, **no se conoce la desviación estándar poblacional** (\(\sigma\)). En estos casos, se utiliza la **desviación estándar muestral** (\(s\)) como estimador de la variabilidad de la población.

Como consecuencia de esta incertidumbre adicional, el cálculo del intervalo de confianza para la media poblacional requiere el uso de la **distribución $t$ de Student** en lugar de la distribución normal estándar ($Z$):

- La distribución $t$ es **similar a la normal**, pero tiene colas más gruesas (es más "plana"), lo que refleja la mayor variabilidad esperada cuando $\sigma$ es desconocida.
- A medida que el tamaño de la muestra aumenta, la distribución $t$ se aproxima a la normal estándar.

### ¿Cuándo usar la distribución $t$?

Se utiliza la distribución $t$ cuando:
- La muestra es pequeña (\(n < 30\)), **y/o**
- La desviación estándar poblacional $\sigma$ es desconocida.

### Fórmula del intervalo de confianza usando $t$ de Student

\[
\text{IC} = \bar{x} \pm t_{\alpha/2,\,df} \cdot \frac{s}{\sqrt{n}}
\]

Donde:
- \(\bar{x}\): media muestral
- \(s\): desviación estándar muestral
- \(n\): tamaño de la muestra
- \(t_{\alpha/2,\,df}\): valor crítico de la distribución $t$ para un nivel de confianza dado y \(df = n-1\) grados de libertad

### **Grados de libertad**

Para la distribución $t$, los **grados de libertad** (\(df\)) se calculan como:

\[
df = n - 1
\]

donde \(n\) es el tamaño de la muestra.

### Resumen

- Si **$\sigma$ es desconocida**, usa $s$ y la distribución $t$.
- Los grados de libertad determinan la forma exacta de la curva $t$.
- Para muestras grandes, la diferencia entre $t$ y $Z$ se vuelve insignificante.

> **Nota:** Siempre verifica la normalidad de los datos, especialmente con muestras pequeñas, ya que la robustez de la distribución $t$ depende de que la población subyacente sea aproximadamente normal.


**Ejemplo**

Un fabricante de cables desea investigar la durabilidad de sus productos. Una muestra de 12 cables para un uso continuo de 10,000 horas reveló una media muestral de 0.48 milímetros de grosor restante con una desviación estándar de 0.12 milímetros. Construya un intervalo de confianza del 95% para la media poblacional. ¿Sería razonable que el fabricante concluyera que después de 10,000 horas la cantidad media poblacional de grosor restante es de 0.45 milímetros?

```python
# Ejemplo: Intervalo de confianza para la media poblacional con σ desconocida

import math
import scipy.stats as stats

# Datos del problema
n = 12
media_muestral = 0.48          # mm de grosor restante
desviacion_muestral = 0.12     # mm
nivel_confianza = 0.95

# Grados de libertad
df = n - 1

# Valor crítico t para el nivel de confianza y grados de libertad
t_alpha_2 = stats.t.ppf(1 - (1 - nivel_confianza) / 2, df)

# Error estándar
error_estandar = desviacion_muestral / math.sqrt(n)

# Margen de error
margen_error = t_alpha_2 * error_estandar

# Intervalo de confianza
limite_inferior = media_muestral - margen_error
limite_superior = media_muestral + margen_error

# Evaluar si 0.45 está dentro del intervalo
conclusion = (0.45 >= limite_inferior) and (0.45 <= limite_superior)

# Mostrar resultados
print(f"Intervalo de confianza al 95%: [{limite_inferior:.3f}, {limite_superior:.3f}] mm")
print(f"¿0.45 mm está dentro del intervalo?: {'Sí' if conclusion else 'No'}")
```

## Intervalos de confianza para una proporción

El **intervalo de confianza para una proporción poblacional** permite estimar, a partir de una muestra, el rango en el que probablemente se encuentra la proporción verdadera de la población con un determinado atributo.

### **Fórmula general**

\[
IC = \hat{p} \pm Z_{\alpha/2} \cdot \sqrt{\frac{\hat{p}(1-\hat{p})}{n}}
\]

Donde:

- \( \hat{p} \): Proporción muestral (número de éxitos dividido entre el tamaño de la muestra).
- \( Z_{\alpha/2} \): Valor crítico de la distribución normal estándar para el nivel de confianza deseado (por ejemplo, para 95%, \( Z_{\alpha/2} \approx 1.96 \)).
- \( n \): Tamaño de la muestra.

### **Pasos para calcular el intervalo de confianza**

1. **Calcular la proporción muestral:**  
   \[
   \hat{p} = \frac{x}{n}
   \]
   donde \( x \) es el número de éxitos en la muestra.

2. **Determinar el valor crítico \( Z_{\alpha/2} \):**  
   - Para un 90% de confianza: \( Z_{\alpha/2} \approx 1.645 \)
   - Para un 95% de confianza: \( Z_{\alpha/2} \approx 1.96 \)
   - Para un 99% de confianza: \( Z_{\alpha/2} \approx 2.576 \)

3. **Calcular el error estándar:**  
   \[
   \text{Error estándar} = \sqrt{\frac{\hat{p}(1-\hat{p})}{n}}
   \]

4. **Construir el intervalo:**  
   \[
   IC = \hat{p} \pm Z_{\alpha/2} \cdot \text{Error estándar}
   \]

### **Condiciones de validez**

- La muestra debe ser aleatoria y representativa.
- Se recomienda que tanto \( n\hat{p} \) como \( n(1-\hat{p}) \) sean mayores o iguales a 5 para que la aproximación normal sea válida.

### **Ejemplo rápido**

Supón que en una muestra de 200 personas, 60 prefieren un producto (\( \hat{p} = 0.30 \)), y deseas un intervalo de confianza del 95%:

1. \( \hat{p} = 0.30 \)
2. \( n = 200 \)
3. \( Z_{0.025} = 1.96 \)
4. Error estándar: \( \sqrt{\frac{0.3 \times 0.7}{200}} \approx 0.032 \)
5. Margen de error: \( 1.96 \times 0.032 \approx 0.063 \)
6. Intervalo: \( 0.30 \pm 0.063 \implies [0.237,\ 0.363] \)

> Con un 95% de confianza, la proporción poblacional está entre 23.7% y 36.3%.

---


```python
# Ejemplo: Intervalo de confianza para una proporción poblacional

import numpy as np
import scipy.stats as stats

# Datos proporcionados
n = 200                # Tamaño de la muestra
x = 60                 # Número de éxitos (personas que prefieren el producto)
nivel_confianza = 0.95

# Proporción muestral
proporcion_muestral = x / n

# Valor crítico Z para el nivel de confianza
Z = stats.norm.ppf(1 - (1 - nivel_confianza) / 2)

# Error estándar de la proporción
error_estandar = np.sqrt((proporcion_muestral * (1 - proporcion_muestral)) / n)

# Intervalo de confianza
limite_inferior = proporcion_muestral - Z * error_estandar
limite_superior = proporcion_muestral + Z * error_estandar

print(f"Proporción muestral: {proporcion_muestral:.3f}")
print(f"Error estándar: {error_estandar:.3f}")
print(f"Valor crítico Z: {Z:.3f}")
print(f"Intervalo de confianza del 95%: [{limite_inferior:.3f}, {limite_superior:.3f}]")
```

```python
# Ejemplo: Proporción e intervalo de confianza para tazas defectuosas

import numpy as np
import scipy.stats as stats

# Datos del problema
n = 300          # Tamaño de la muestra
x = 15           # Número de tazas defectuosas
nivel_confianza = 0.95

# a) Proporción muestral de tazas defectuosas
proporcion_muestral = x / n

# b) Intervalo de confianza del 95% para la proporción
Z = stats.norm.ppf(1 - (1 - nivel_confianza) / 2)
error_estandar = np.sqrt((proporcion_muestral * (1 - proporcion_muestral)) / n)
limite_inferior = proporcion_muestral - Z * error_estandar
limite_superior = proporcion_muestral + Z * error_estandar

# c) Decisión sobre devolución del lote
umbral_devolucion = 0.10
devolver = limite_inferior >= umbral_devolucion

# Mostrar resultados
print(f"Proporción muestral de tazas defectuosas: {proporcion_muestral:.3f}")
print(f"Error estándar: {error_estandar:.3f}")
print(f"Intervalo de confianza del 95%: [{limite_inferior:.3f}, {limite_superior:.3f}]")
if devolver:
    print("Decisión: Zack debe devolver el lote (el 10% o más de tazas podrían ser defectuosas).")
else:
    print("Decisión: No es necesario devolver el lote (la proporción de defectuosas es menor al 10%).")
```

## Factor de corrección de una población finita

El **factor de corrección de una población finita** (FC) es esencial cuando se trabaja con muestras extraídas de poblaciones finitas, especialmente si la muestra representa una fracción considerable de la población total. Este factor ajusta el cálculo del error estándar y la varianza, haciendo que las estimaciones sean más precisas y representativas.

---

### **¿Por qué es importante?**

- **Mejora la precisión:** Ajusta el error estándar, afinando la estimación de parámetros como la media o la proporción poblacional.
- **Evita sobreestimar el error:** Sin el factor de corrección, el error muestral puede ser exagerado, resultando en intervalos de confianza innecesariamente amplios.
- **Relevancia según el tamaño muestral:** Es especialmente importante cuando la muestra supera el 5-10% del tamaño de la población.

---

### **Fórmula del factor de corrección**

\[
FC = \sqrt{\frac{N - n}{N - 1}}
\]

Donde:
- \( N \): tamaño de la población
- \( n \): tamaño de la muestra

---

### **¿Cuándo debe aplicarse?**

- Cuando la población es finita y la muestra es significativa respecto al total.
- Si la muestra representa más del 5-10% de la población, el uso del factor de corrección es recomendable.

---

### **Aplicación al error estándar**

El error estándar ajustado para una población finita se calcula así:

\[
SE_{ajustado} = SE \times FC
\]

Donde:
- \( SE \): error estándar sin ajustar
- \( SE_{ajustado} \): error estándar ajustado con el factor de corrección

---

### **Ejemplo práctico**

Supón una población de **5,000 estudiantes** y una muestra aleatoria de **300** estudiantes. Queremos estimar la proporción de quienes prefieren estudiar en la biblioteca.

#### **Paso 1: Proporción muestral**

\[
\hat{p} = \frac{180}{300} = 0.6
\]

#### **Paso 2: Error estándar sin corrección**

\[
SE = \sqrt{\frac{\hat{p}(1 - \hat{p})}{n}} = \sqrt{\frac{0.6 \times 0.4}{300}} \approx 0.0283
\]

#### **Paso 3: Factor de corrección**

\[
FC = \sqrt{\frac{5000 - 300}{5000 - 1}} \approx \sqrt{\frac{4700}{4999}} \approx 0.969
\]

#### **Paso 4: Error estándar ajustado**

\[
SE_{ajustado} = 0.0283 \times 0.969 \approx 0.0274
\]

#### **Paso 5: Intervalo de confianza del 95%**

\[
IC = 0.6 \pm 1.96 \times 0.0274 \implies [0.546,\, 0.654]
\]

---

### **Resumen**

- El factor de corrección de una población finita ajusta el error estándar para muestras grandes en poblaciones finitas.
- Su uso proporciona **intervalos de confianza más precisos** y evita conclusiones erróneas sobre la población.
- Es fundamental cuando la muestra es una fracción considerable del total poblacional.

> **Recomendación:** Aplica siempre el factor de corrección cuando la muestra supera el 5-10% de la población para obtener inferencias más fiables.


```python
# Ejemplo: Intervalo de confianza para una proporción con corrección para población finita

import math
import scipy.stats as stats

# Datos del ejemplo
N = 5000    # Tamaño de la población
n = 300     # Tamaño de la muestra
x = 180     # Número de estudiantes que prefieren estudiar en la biblioteca

# Proporción muestral
p_hat = x / n

# Error estándar sin corrección
SE = math.sqrt(p_hat * (1 - p_hat) / n)

# Factor de corrección para población finita
FC = math.sqrt((N - n) / (N - 1))

# Error estándar ajustado
SE_ajustado = SE * FC

# Valor crítico Z para un nivel de confianza del 95%
Z = stats.norm.ppf(1 - 0.05 / 2)

# Intervalo de confianza ajustado
limite_inferior = p_hat - Z * SE_ajustado
limite_superior = p_hat + Z * SE_ajustado

# Mostrar resultados
print(f"Proporción muestral: {p_hat:.3f}")
print(f"Error estándar sin corrección: {SE:.4f}")
print(f"Factor de corrección: {FC:.3f}")
print(f"Error estándar ajustado: {SE_ajustado:.4f}")
print(f"Valor crítico Z: {Z:.3f}")
print(f"Intervalo de confianza del 95%: [{limite_inferior:.3f}, {limite_superior:.3f}]")
```

## Elección y Cálculo del Tamaño de la Muestra

La **elección del tamaño de muestra** es un paso clave para garantizar resultados precisos y representativos en cualquier estudio estadístico. Un tamaño de muestra adecuado reduce el error de muestreo y aumenta la confianza en las inferencias sobre la población.

---

### **Factores a Considerar**

1. **Nivel de confianza (\( Z \))**
   - Refleja la seguridad con la que se desea estimar el parámetro poblacional.
   - Valores comunes: 90% (\(Z \approx 1.645\)), 95% (\(Z \approx 1.96\)), 99% (\(Z \approx 2.576\)).

2. **Error de muestreo tolerable (\( E \))**
   - Es la máxima diferencia aceptable entre el estimador muestral y el parámetro poblacional.
   - A menor \( E \), mayor tamaño de muestra requerido.

3. **Proporción poblacional (\( p \))**
   - Es la fracción estimada de la población con la característica de interés.
   - Si se desconoce, se usa \( p = 0.5 \) para obtener el tamaño de muestra más conservador.

4. **Variabilidad de la población (\( \sigma \))**
   - Es la desviación estándar de la población. Si no se conoce, puede estimarse con datos previos o una muestra piloto.

---

### **Fórmulas para el Tamaño de la Muestra**

#### a) Para una proporción:

\[
n = \frac{Z^2 \cdot p \cdot (1 - p)}{E^2}
\]

#### b) Para una media:

\[
n = \left( \frac{Z \cdot \sigma}{E} \right)^2
\]

---

### **Ejemplo 1: Tamaño de muestra para una proporción**

Supón que deseas estimar una proporción poblacional con:
- Nivel de confianza: 95% (\(Z = 1.96\))
- Error tolerable: 5% (\(E = 0.05\))
- Proporción estimada: \(p = 0.5\)

\[
n = \frac{1.96^2 \cdot 0.5 \cdot 0.5}{0.05^2} = \frac{3.8416 \cdot 0.25}{0.0025} = \frac{0.9604}{0.0025} = 384.16
\]

**Resultado:**  
Se recomienda una muestra de al menos **385 individuos** para estimar la proporción con un 95% de confianza y un error de ±5%.

---

### **Ejemplo 2: Tamaño de muestra para una media**

Un investigador desea estimar el gasto medio mensual en entretenimiento de los hogares de una ciudad, con:
- Nivel de confianza: 95% (\(Z = 1.96\))
- Error tolerable: \$50 (\(E = 50\))
- Desviación estándar: \$500 (\(\sigma = 500\))

\[
n = \left( \frac{1.96 \cdot 500}{50} \right)^2 = \left( \frac{980}{50} \right)^2 = (19.6)^2 = 384.16
\]

**Resultado:**  
Se requiere una muestra de al menos **385 hogares** para cumplir con los criterios establecidos.

---

### **Ejemplo 3: Proporción de turistas en Las Vegas**

- Proporción estimada: \(p = 0.3\)
- Nivel de confianza: 90% (\(Z = 1.645\))
- Error tolerable: 1% (\(E = 0.01\))

\[
n = \frac{1.645^2 \cdot 0.3 \cdot 0.7}{0.01^2} = \frac{2.706 \cdot 0.21}{0.0001} = \frac{0.56826}{0.0001} = 5682.6
\]

**a) Tamaño de muestra necesario:**  
Se necesitaría una muestra de al menos **5,683 turistas** para estimar la proporción con un error de ±1% y 90% de confianza.

---

**b) ¿Cómo reducir el tamaño de la muestra?**

- **Aumentar el error tolerable (\(E\))**: Permitir un margen de error mayor reduce el tamaño de muestra.
- **Disminuir el nivel de confianza**: Usar un nivel de confianza menor (por ejemplo, 85% en vez de 90%).
- **Utilizar información previa más precisa sobre \(p\)**: Si se tiene una mejor estimación de la proporción, el tamaño de muestra puede disminuir.

**Ejemplo con error tolerable de 2% (\(E = 0.02\)):**

\[
n = \frac{1.645^2 \cdot 0.3 \cdot 0.7}{0.02^2} = \frac{0.56826}{0.0004} = 1420.7
\]

**Nuevo tamaño de muestra:**  
Con un error tolerable de 2%, se requiere una muestra de **1,421 turistas**.

---

### **Conclusiones y Recomendaciones**

- El tamaño de muestra depende del nivel de confianza, el error tolerable, la variabilidad y la proporción estimada.
- Para muestras grandes respecto a la población total, considera aplicar el **factor de corrección para población finita**.
- Siempre justifica y documenta las decisiones tomadas sobre el tamaño de muestra, ya que afectan la validez de los resultados.

> **Recuerda:** Un tamaño de muestra adecuado es clave para obtener resultados estadísticos confiables y útiles para la toma de decisiones.


```python
# Ejemplo 1: Tamaño de muestra para una proporción

import math
import scipy.stats as stats

# Parámetros del ejemplo
nivel_confianza = 0.95
Z = stats.norm.ppf(1 - (1 - nivel_confianza) / 2)
p = 0.5          # Proporción estimada
E = 0.05         # Error tolerable

# Fórmula para el tamaño de muestra para una proporción
n_proporcion = (Z**2 * p * (1 - p)) / (E**2)
print(f"Tamaño de muestra para proporción (95% confianza, error ±5%): {math.ceil(n_proporcion)}")
```

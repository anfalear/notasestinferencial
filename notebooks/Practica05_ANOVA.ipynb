### Análisis de la varianza (ANOVA de una vía)

El **Análisis de la Varianza (ANOVA) de una vía** es una técnica estadística diseñada para determinar si existen diferencias estadísticamente significativas entre las medias de tres o más grupos independientes[1][3][4][6]. Se utiliza cuando se cuenta con una **variable dependiente cuantitativa** y una **variable independiente categórica** (factor) que tiene al menos tres niveles o categorías[1][7]. Por ejemplo, puede emplearse para comparar el rendimiento académico de estudiantes que utilizan diferentes métodos de estudio (A, B y C) o para analizar la eficacia de distintos tratamientos médicos[3][4][6].

#### Hipótesis

- **Hipótesis nula ($$H_0$$)**: Todas las medias poblacionales de los grupos son iguales ($$\mu_1 = \mu_2 = \cdots = \mu_k$$), lo que implica que cualquier diferencia observada entre los grupos es atribuible únicamente al azar[1][3][4].
- **Hipótesis alternativa ($$H_1$$)**: Al menos una de las medias es distinta. No todas las medias son iguales; es decir, existe al menos un grupo cuya media es significativamente diferente del resto[1][3][4].

#### Concepto clave: Variabilidad

El ANOVA mide y compara la variabilidad en los datos de dos formas fundamentales:

- **Variabilidad entre grupos (SSB)**: Corresponde a las diferencias entre las medias de los grupos y la media general. Una SSB alta sugiere que los grupos podrían ser diferentes[1][5][6].
- **Variabilidad dentro de los grupos (SSW)**: Refiere a la dispersión de los datos individuales respecto a su propia media de grupo. Una SSW alta indica mayor dispersión interna y menos evidencia de diferencias reales entre grupos[1][5][6].

El fundamento del ANOVA es comparar la varianza explicada por la agrupación (entre grupos) frente a la varianza no explicada (dentro de grupos)[5].

#### Estadístico F

El **estadístico F** se utiliza en ANOVA para determinar si la variabilidad explicada por el modelo (entre grupos) es significativamente mayor que la variabilidad no explicada (dentro de grupos):

$$
F = \frac{\text{MSB}}{\text{MSW}}
$$

- **MSB (Mean Square Between)**: Media cuadrática entre los grupos, calculada como SSB dividido por los grados de libertad entre grupos ($$\text{df}_1 = k-1$$).
- **MSW (Mean Square Within)**: Media cuadrática dentro de los grupos, calculada como SSW dividido por los grados de libertad dentro de grupos ($$\text{df}_2 = N-k$$), donde N es el total de observaciones.

**Interpretación**:
- Si $$F$$ es considerablemente grande y el valor-p asociado es menor que el nivel de significancia (habitualmente 0.05), se rechaza la hipótesis nula, concluyendo que existen diferencias significativas entre las medias de los grupos[1][4][6].
- Si $$F$$ es pequeño, no hay evidencia suficiente para afirmar que las medias de los grupos difieran estadísticamente.

#### Supuestos del ANOVA de una vía

Para que los resultados del ANOVA sean válidos, se deben cumplir los siguientes supuestos[1][6][7]:

- **Independencia de las observaciones:** Los datos de cada grupo son independientes entre sí.
- **Normalidad:** Las distribuciones de los datos dentro de cada grupo deben ser aproximadamente normales (verificable, por ejemplo, con la prueba Shapiro-Wilk).
- **Homogeneidad de varianzas:** Las varianzas poblacionales de los grupos son iguales (puede comprobarse con la prueba de Levene).

Si estos supuestos no se cumplen, es recomendable usar pruebas no paramétricas como **Kruskal-Wallis**[7].

#### Proceso para realizar un ANOVA de una vía

1. **Revisión de supuestos:** Comprobar normalidad de los datos, homogeneidad de varianzas y la independencia de las observaciones[1][6][7].
2. **Cálculo de SSB, SSW y estadístico F:** Determinar las sumas de cuadrados, medias cuadráticas y el estadístico F.
3. **Cómputo e interpretación del valor-p:** Si el valor-p es inferior al umbral de significancia ($$\alpha = 0.05$$), se rechaza la hipótesis nula.
4. **Pruebas post-hoc:** En caso de encontrar diferencias significativas, aplicar pruebas post-hoc (como Tukey o Bonferroni) para identificar específicamente cuáles grupos difieren entre sí[6].

#### Ejemplo práctico de ANOVA

Suponga que se compara el efecto de tres fertilizantes (X, Y, Z) sobre el crecimiento de plantas. El ANOVA arroja un estadístico $$ F = 8.4 $$ y un valor-p de 0.001. Dado que el valor-p es menor a 0.05, se rechaza la hipótesis nula, concluyendo que al menos uno de los fertilizantes produce un efecto diferente en el crecimiento. Un análisis post-hoc podría revelar, por ejemplo, que el fertilizante Y es significativamente mejor que X y Z[3][6].

#### ¿Por qué no usar múltiples pruebas t?

Realizar comparaciones por pares con múltiples t-tests aumenta la probabilidad de cometer **errores Tipo I** (encontrar diferencias donde no las hay). El ANOVA controla este error, permitiendo analizar todos los grupos simultáneamente y sólo indicar diferencias reales cuando existen[1][5][6].

#### Comprensión de la distribución F

La **distribución F** es una distribución de probabilidad continua y asimétrica (solo toma valores positivos), utilizada para comparar la varianza explicada por el modelo con la no explicada. En el ANOVA, se emplea para determinar la significación estadística del estadístico F, considerando los grados de libertad asociados al numerador (entre grupos) y al denominador (dentro de grupos)[6].

#### Aplicaciones comunes del ANOVA

- Comparación de medias en diferentes grupos experimentales o tratamientos[4][5].
- Evaluación de la productividad o calidad bajo diferentes condiciones en procesos industriales.
- Estudios educativos, médicos o sociales para determinar la efectividad de métodos, terapias o intervenciones[4][5][6].

#### Observaciones adicionales

- Si los supuestos de normalidad y homogeneidad de varianzas no se cumplen, recurra a alternativas no paramétricas[7].
- La interpretación de los resultados suele apoyarse con representaciones gráficas (por ejemplo, diagramas de caja), que ayudan a identificar tendencias y valores atípicos[1].
- El ANOVA no identifica por sí mismo qué grupos difieren, sino solo si existe o no diferencia general; por tanto, al obtener un resultado significativo, es imprescindible realizar pruebas post-hoc para comparaciones múltiples[6].

#### Resumen visual y ejemplo aplicado

Ejemplo: Se analiza el desempeño (sobre 100) de estudiantes divididos en tres métodos de enseñanza:

- Método A: [85, 88, 82, B: [75, 78,o C: [80, 85,ener un valor $$ F = 14.03 $$ (df₁ = 2, df₂ = 12, valor-p = 0.0006), se concluye que existen diferencias significativas en las medias. Una prueba post-hoc indica que los métodos A y C superan al B, sin diferencia significativa entre A y C.

El ANOVA de una vía, desarrollado originalmente por Ronald A. Fisher, es un pilar de la inferencia estadística para la toma de decisiones y la validación de hipótesis al comparar múltiples grupos a partir de una variable respuesta continua[6].

La empresa Lammers Limos ofrece un servicio de transporte en limusina que conecta el ayuntamiento de Toledo, Ohio, con el aeropuerto metropolitano de Detroit. Sean Lammers, presidente de la compañía, está considerando dos posibles rutas para brindar este servicio: una que utiliza la carretera 25 y otra que recorre la autopista I-75. Con el objetivo de evaluar cuál ruta es más conveniente en términos de consistencia en los tiempos de viaje, Lammers recopiló datos muestrales sobre la duración, en minutos, de varios recorridos realizados por ambas rutas. Los tiempos registrados para la carretera 25 fueron 52, 67, 56, 45, 71, 54 y 64 minutos, mientras que para la autopista I-75 los tiempos obtenidos fueron 59, 60, 61, 51, 56, 63, 57 y 65 minutos. Utilizando un nivel de significancia de 0.10, se plantea la pregunta de si existe una diferencia significativa en la variabilidad de los tiempos de manejo entre estas dos alternativas de ruta.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Datos de los tiempos (en minutos) por ruta
carretera_25 = [52, 67, 56, 45, 71, 54, 64]
autopista_i75 = [59, 60, 61, 51, 56, 63, 57, 65]

def comparar_varianzas(grupo1, grupo2, alpha=0.10):
    # Prueba de normalidad (Shapiro-Wilk)
    _, p_shapiro1 = stats.shapiro(grupo1)
    _, p_shapiro2 = stats.shapiro(grupo2)

    # Varianzas muestrales (ddof=1)
    var1 = np.var(grupo1, ddof=1)
    var2 = np.var(grupo2, ddof=1)

    # Determinar qué varianza es mayor para el cálculo F correcto
    if var1 > var2:
        f_ratio = var1 / var2
        dfn = len(grupo1) - 1  # grados de libertad del numerador
        dfd = len(grupo2) - 1  # grados de libertad del denominador
    else:
        f_ratio = var2 / var1
        dfn = len(grupo2) - 1
        dfd = len(grupo1) - 1

    # Prueba F bilateral para igualdad de varianzas
    p_value = 2 * (1 - stats.f.cdf(f_ratio, dfn, dfd))

    # Prueba de Levene (robusta ante no normalidad)
    _, p_levene = stats.levene(grupo1, grupo2)

    print("--------------------------------------")
    print("          Resultados clave")
    print("--------------------------------------")
    print(f"Varianza Carretera 25: {var1:.2f}")
    print(f"Varianza Autopista I-75: {var2:.2f}")
    print(f"F-estadístico: {f_ratio:.2f}")
    print(f"Valor-p (prueba F): {p_value:.4f}")
    print(f"Valor-p (Shapiro-Wilk - Carretera 25): {p_shapiro1:.4f}")
    print(f"Valor-p (Shapiro-Wilk - Autopista I-75): {p_shapiro2:.4f}")
    print(f"Valor-p (Levene): {p_levene:.4f}")

    # Interpretación de resultados
    print("\n--------------------------------------")
    print("           Interpretación")
    print("--------------------------------------")
    if p_shapiro1 > alpha and p_shapiro2 > alpha:
        print("Supuesto de normalidad: OK (p > 0.10)")
        print("Conclusión basada en la prueba F:")
        if p_value < alpha:
            print("✅ Rechazamos H₀: Hay diferencia significativa en varianzas.")
        else:
            print("❌ No hay evidencia para rechazar H₀: Varianzas similares.")
    else:
        print("Supuesto de normalidad: Violado (p ≤ 0.10)")
        print("Conclusión basada en la prueba de Levene (más robusta):")
        if p_levene < alpha:
            print("✅ Rechazamos H₀: Hay diferencia significativa en varianzas.")
        else:
            print("❌ No hay evidencia para rechazar H₀: Varianzas similares.")

# Ejecutar análisis estadístico
comparar_varianzas(carretera_25, autopista_i75, alpha=0.10)

# Gráfica de boxplot para comparación visual
plt.figure(figsize=(8, 5))
plt.boxplot([carretera_25, autopista_i75], labels=["Carretera 25", "Autopista I-75"])
plt.title("Distribución de Tiempos por Ruta")
plt.ylabel("Minutos")
plt.grid(True)
plt.show()
```

### Evaluación del Nivel de Satisfacción en Cuatro Aerolíneas

En los últimos años, es común observar que las aerolíneas han reducido ciertos servicios que antes ofrecían durante sus vuelos, como alimentos y refrigerios gratuitos, e incluso han comenzado a cobrar por servicios adicionales tales como transportar equipaje con sobrepeso, realizar cambios de vuelo de último momento o permitir mascotas en la cabina. No obstante, las compañías continúan interesadas en la percepción que los pasajeros tienen acerca de su servicio.

Recientemente, un grupo conformado por cuatro aerolíneas (designadas aquí con nombres históricos por cuestiones de confidencialidad) contrató a Brunner Marketing Research, Inc. para realizar una encuesta de satisfacción a sus pasajeros acerca de distintos aspectos como la compra de boletos, el proceso de abordaje, la atención recibida durante el vuelo, el manejo del equipaje y la comunicación del piloto, entre otros.

La encuesta consistió en 25 preguntas con opciones de respuesta: excelente (calificación de 4), bueno (3), regular (2) y deficiente (1). Las respuestas se sumaron para obtener una calificación total, donde un mayor puntaje indica mayor satisfacción con el servicio recibido, siendo 100 la mayor calificación posible.

De manera aleatoria, se seleccionaron y evaluaron pasajeros de cada una de las cuatro aerolíneas; las puntuaciones obtenidas fueron las siguientes:

- **Eastern:** 94, 90, 85, 80
- **TWA:** 75, 68, 77, 83, 88
- **Allegheny:** 70, 73, 76, 78, 80, 68, 65
- **Ozark:** 68, 70, 72, 65, 74, 65

**Pregunta:**  
¿Existe alguna diferencia significativa en el nivel promedio de satisfacción entre las cuatro aerolíneas? Utilice un nivel de significancia de 0.01 para su análisis.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
from statsmodels.stats.multicomp import pairwise_tukeyhsd
from scipy.stats import shapiro, levene

# Paso 1: Datos y revisión de supuestos

groups = {
    "Eastern": [94, 90, 85, 80],
    "TWA": [75, 68, 77, 83, 88],
    "Allegheny": [70, 73, 76, 78, 80, 68, 65],
    "Ozark": [68, 70, 72, 65, 74, 65]
}

alpha = 0.05  # Nivel de significancia

# Normalidad para cada grupo (Shapiro-Wilk)
for name, data in groups.items():
    stat, p = shapiro(data)
    print(f"Normalidad (Shapiro-Wilk) {name}: stat={stat:.3f}, p={p:.4f}")

# Homogeneidad de varianzas (Levene)
stat_levene, p_levene = levene(*(v for v in groups.values()))
print(f"Homogeneidad de varianzas (Levene): stat={stat_levene:.3f}, p={p_levene:.4f}")

# Paso 2: ANOVA (SSB, SSW y estadístico F)

all_data = np.concatenate(list(groups.values()))
group_labels = np.concatenate([[k] * len(v) for k, v in groups.items()])
means = {k: np.mean(v) for k, v in groups.items()}
mean_global = np.mean(all_data)

# Suma de cuadrados entre grupos (SSB)
ss_between = sum(len(groups[k]) * (means[k] - mean_global)**2 for k in groups)
# Suma de cuadrados dentro de los grupos (SSW)
ss_within = sum(sum((x - means[k])**2 for x in groups[k]) for k in groups)

k = len(groups)
n_total = len(all_data)
df_between = k - 1
df_within = n_total - k
ms_between = ss_between / df_between
ms_within = ss_within / df_within

# Estadístico F y valor-p
F_stat = ms_between / ms_within
p_value = 1 - stats.f.cdf(F_stat, df_between, df_within)

print("\nTabla ANOVA")
print(f"{'Fuente':<15}{'SS':>10}{'df':>6}{'MS':>10}{'F':>10}{'valor-p':>12}")
print(f"{'Entre grupos':<15}{ss_between:10.2f}{df_between:6}{ms_between:10.2f}{F_stat:10.2f}{p_value:12.4f}")
print(f"{'Dentro grupos':<15}{ss_within:10.2f}{df_within:6}{ms_within:10.2f}{'':>10}{'':>12}")
print(f"{'Total':<15}{(ss_between+ss_within):10.2f}{df_between+df_within:6}")

# Paso 3: Interpretación
if p_value < alpha:
    print("\n→ Hay diferencias significativas entre los niveles de satisfacción (p < 0.05)")
else:
    print("\n→ No hay diferencias significativas entre los niveles de satisfacción (p >= 0.05)")

# Paso 4: Pruebas post-hoc (Tukey HSD) si corresponde
if p_value < alpha:
    tukey_result = pairwise_tukeyhsd(endog=all_data, groups=group_labels, alpha=alpha)
    print("\nResultados de la prueba post-hoc de Tukey:")
    print(tukey_result)

# Intervalos de confianza (ejemplo al 99%) para cada grupo
confidence = 0.99
print("\nIntervalos de confianza del 99% para la media de cada grupo:")
for name, data in groups.items():
    n = len(data)
    mean = np.mean(data)
    std_err = np.std(data, ddof=1) / np.sqrt(n)
    t_crit = stats.t.ppf(1 - (1-confidence)/2, df=n-1)
    margin = t_crit * std_err
    print(f"{name}: Media = {mean:.2f}, IC = [{mean-margin:.2f}, {mean+margin:.2f}]")

# Gráfica: intervalos de confianza
plt.figure(figsize=(8, 6))
for idx, (name, data) in enumerate(groups.items()):
    n = len(data)
    mean = np.mean(data)
    std_err = np.std(data, ddof=1) / np.sqrt(n)
    t_crit = stats.t.ppf(1 - (1-confidence)/2, df=n-1)
    margin = t_crit * std_err
    plt.plot([idx, idx], [mean-margin, mean+margin], color='blue', marker='o')
    plt.scatter(idx, mean, color='red')
plt.xticks(range(len(groups)), list(groups.keys()), rotation=45)
plt.ylabel("Nivel de satisfacción")
plt.title("Intervalos de Confianza de las Aerolíneas")
plt.grid(True)
plt.show()
```

Una empresa administradora de centros comerciales está interesada en comparar el desempeño financiero de sus locatarios en tres de sus principales ubicaciones: **Southwyck**, **Franklin Park** y **Old Orchard**. Para ello, recopiló información sobre los **ingresos mensuales promedio** (en miles de dólares) de cuatro tiendas seleccionadas al azar en cada centro comercial. El objetivo del estudio es determinar si existen diferencias significativas en los ingresos entre estas ubicaciones, lo cual podría influir en decisiones estratégicas como la asignación de recursos publicitarios, la fijación de rentas o la planificación de remodelaciones.

**Problema:**  
¿Existen diferencias estadísticamente significativas en los ingresos promedio entre los tres centros comerciales? Utilice un análisis adecuado (como un ANOVA de una vía) para evaluar si, al menos, uno de los centros comerciales muestra un nivel de ingresos distinto respecto a los demás.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
from statsmodels.stats.multicomp import pairwise_tukeyhsd
from scipy.stats import shapiro, levene

# Datos de ingresos en miles de dólares para cada área
groups = {
    "Southwyck": np.array([64, 68, 70, 60]),
    "Franklin Park": np.array([74, 71, 69, 70]),
    "Old Orchard": np.array([75, 80, 76, 78])
}
industrias = list(groups.keys())
alpha = 0.05

# Paso 1: Supuestos estadísticos
print("\n--- SUPUESTOS ESTADÍSTICOS ---")
for name, data in groups.items():
    stat, p = shapiro(data)
    print(f"Normalidad (Shapiro-Wilk) {name}: stat={stat:.3f}, p={p:.4f}")
stat_levene, p_levene = levene(*(v for v in groups.values()))
print(f"Homogeneidad de varianzas (Levene): stat={stat_levene:.3f}, p={p_levene:.4f}")

# Paso 2: ANOVA de un factor (manual)
all_data = np.concatenate(list(groups.values()))
group_labels = np.concatenate([[k] * len(v) for k, v in groups.items()])
means = {k: np.mean(v) for k, v in groups.items()}
mean_global = np.mean(all_data)
k = len(groups)
n_total = len(all_data)
df_between = k - 1
df_within = n_total - k
ss_between = sum(len(groups[k]) * (means[k] - mean_global)**2 for k in groups)
ss_within = sum(sum((x - means[k])**2 for x in groups[k]) for k in groups)
ms_between = ss_between / df_between
ms_within = ss_within / df_within
F_stat = ms_between / ms_within
p_value = 1 - stats.f.cdf(F_stat, df_between, df_within)

print("\n--- TABLA ANOVA ---")
print(f"{'Fuente':<15}{'SS':>10}{'df':>6}{'MS':>10}{'F':>10}{'valor-p':>12}")
print(f"{'Entre grupos':<15}{ss_between:10.2f}{df_between:6}{ms_between:10.2f}{F_stat:10.2f}{p_value:12.4f}")
print(f"{'Dentro grupos':<15}{ss_within:10.2f}{df_within:6}{ms_within:10.2f}{'':>10}{'':>12}")
print(f"{'Total':<15}{(ss_between+ss_within):10.2f}{df_between+df_within:6}")

# Paso 3: Interpretación
if p_value < alpha:
    print("\n→ Hay diferencias significativas entre ingresos medios de las áreas (p < 0.05)")
else:
    print("\n→ No hay diferencias significativas entre ingresos medios de las áreas (p >= 0.05)")

# Paso 4: Prueba post-hoc de Tukey HSD (si aplica)
if p_value < alpha:
    tukey_result = pairwise_tukeyhsd(endog=all_data, groups=group_labels, alpha=alpha)
    print("\n--- PRUEBA POST-HOC DE TUKEY ---")
    print(tukey_result)

    # Gráfico de Tukey
    fig, ax = plt.subplots(figsize=(8, 5))
    tukey_result.plot_simultaneous(ax=ax, ylabel="Área", xlabel="Diferencia de medias")
    plt.title("Resultados de la Prueba de Tukey HSD")
    plt.grid(True, linestyle="--", alpha=0.6)
    plt.show()

# Cálculo y gráfico de intervalos de confianza (95%)
def intervalo_confianza(datos):
    media = np.mean(datos)
    std_error = np.std(datos, ddof=1) / np.sqrt(len(datos))
    t_value = stats.t.ppf(1 - alpha / 2, df=len(datos) - 1)
    intervalo = t_value * std_error
    return media, media - intervalo, media + intervalo

medias, ic_inf, ic_sup = [], [], []
for name in industrias:
    media, inf, sup = intervalo_confianza(groups[name])
    medias.append(media)
    ic_inf.append(inf)
    ic_sup.append(sup)
print("\n--- INTERVALOS DE CONFIANZA (95%) ---")
for i, name in enumerate(industrias):
    print(f"{name}: Media = {medias[i]:.2f}, IC = [{ic_inf[i]:.2f}, {ic_sup[i]:.2f}]")

plt.figure(figsize=(8, 5))
plt.bar(industrias, medias, yerr=[np.array(medias) - np.array(ic_inf), np.array(ic_sup) - np.array(medias)],
        capsize=5, color=["blue", "orange", "green"], alpha=0.7)
plt.xlabel("Área")
plt.ylabel("Ingreso medio ($1000)")
plt.title("Intervalos de confianza (95%) para cada área")
plt.grid(axis="y", linestyle="--", alpha=0.5)
plt.tight_layout()
plt.show()
```

### ANOVA de dos vías

El **Análisis de la Varianza de dos vías** (ANOVA de dos factores) es una técnica estadística avanzada que permite evaluar el efecto simultáneo de dos variables categóricas (factores) sobre una variable dependiente cuantitativa. Además de analizar el efecto individual (principal) de cada factor, también es capaz de detectar la existencia de una **interacción** entre ambos, es decir, si el efecto de un factor depende del nivel del otro.

#### ¿Cuándo se utiliza ANOVA de dos vías?

Se emplea cuando se desea comparar medias de más de dos grupos generados en función de **dos factores categóricos independientes**. Esto permite:

- Identificar si existen diferencias en las medias debido a cada factor (efectos principales).
- Determinar si los efectos de un factor sobre la variable respuesta son diferentes según los niveles del segundo factor (efecto de interacción).

**Ejemplo típico:**  
Un estudio de salarios en una empresa, donde se desea analizar simultáneamente el efecto del área de trabajo (Manufactura, Marketing, Investigación y Desarrollo) y el género (Femenino, Masculino) sobre el salario mensual.

#### Formulación de Hipótesis

Para cada factor y para la interacción se plantean hipótesis:

- **Efecto principal del Factor A:**  
  - $$ H_{0A}: $$ Las medias son iguales entre los niveles de A.  
  - $$ H_{1A}: $$ Al menos una media difiere entre los niveles de A.
- **Efecto principal del Factor B:**  
  - $$ H_{0B}: $$ Las medias son iguales entre los niveles de B.  
  - $$ H_{1B}: $$ Al menos una media difiere entre los niveles de B.
- **Interacción A × B:**  
  - $$ H_{0AB}: $$ No existe interacción entre A y B (los efectos son aditivos).  
  - $$ H_{1AB}: $$ Existe interacción (el efecto de un factor depende del nivel del otro).

#### Concepto Clave: Variabilidad y Particiones

El ANOVA de dos vías descompone la variabilidad total de la variable dependiente en cuatro componentes principales:

1. **Variabilidad debida al Factor A**
2. **Variabilidad debida al Factor B**
3. **Variabilidad debida a la interacción A × B**
4. **Variabilidad residual o error**

Cada una de estas fuentes se asocia a sumas de cuadrados (SS), medias cuadráticas (MS) y sus respectivos grados de libertad.

#### Estadístico F en ANOVA de dos vías

Para cada fuente de variabilidad, se calcula un estadístico F independiente:

$$
F_{A} = \frac{\text{MS}_{A}}{\text{MS}_{\text{Error}}} \qquad
F_{B} = \frac{\text{MS}_{B}}{\text{MS}_{\text{Error}}} \qquad
F_{AB} = \frac{\text{MS}_{AB}}{\text{MS}_{\text{Error}}}
$$

- **MS:** Media cuadrática de cada fuente ($$\text{SS}/\text{df}$$).
- **Error:** Corresponde a la variabilidad no explicada por los factores o su interacción.
- Si el valor F (y el valor-p asociado) para un factor es significativo, se concluye que dicho factor (o su interacción) tiene efecto sobre la variable dependiente.

#### Supuestos de ANOVA de dos vías

- **Independencia de observaciones**
- **Normalidad** de la variable respuesta en cada combinación de factores (verificable con pruebas como Shapiro-Wilk)
- **Homogeneidad de varianzas** en todos los grupos combinados (verificable con la prueba de Levene)

Si los supuestos no se cumplen, pueden emplearse alternativas como ANOVA no paramétrico para dos factores.

#### Proceso para realizar un ANOVA de dos vías

1. **Verificar los supuestos** de validez.
2. **Identificar los factores, niveles y la variable dependiente.**
3. **Calcular las sumas de cuadrados y medias cuadráticas** para cada fuente (Factor A, Factor B, Interacción, Error).
4. **Obtener los estadísticos F** y sus respectivos valores-p.
5. **Interpretar resultados:**  
   - Si existe interacción significativa, se recomienda analizar dicha interacción antes de interpretar los efectos principales.
6. **Realizar comparaciones post-hoc** si algún efecto principal es significativo.

#### Ejemplo Práctico

Suponga una empresa que desea analizar si los salarios mensuales de sus empleados varían según el área (Manufactura, Marketing, Investigación y Desarrollo) y el género (Femenino, Masculino). Se recolectan los salarios de varios empleados para cada combinación de área y género.

- Se aplica un **ANOVA de dos vías** para determinar si existen diferencias salariales atribuibles al área de trabajo, al género o a la interacción entre ambos factores.
- Si, por ejemplo, se detecta una interacción significativa, esto podría indicar que la diferencia salarial entre hombres y mujeres varía según el área.

#### Ventajas del ANOVA de dos vías

- Permite evaluar de forma eficiente el impacto de dos factores y su interacción con un solo análisis.
- Reduce la probabilidad de errores Tipo I frente a realizar múltiples tests independientes.
- Ofrece una visión más completa en estudios multifactoriales, como los organizacionales, industriales, biomédicos y educativos.

#### Notas finales

- Si los supuestos no se cumplen, considere métodos alternativos (como la prueba de Friedman para datos no paramétricos o transformaciones de datos).
- Es útil completar el análisis con representaciones visuales (diagramas de interacción o boxplots) para interpretar correctamente los resultados, especialmente en presencia de interacción significativa.
- El análisis post-hoc puede realizarse en cualquiera de los factores si los efectos principales resultan significativos.

El ANOVA de dos vías es una herramienta indispensable para el análisis factorial y la investigación de efectos complejos sobre variables cuantitativas en estudios experimentales y observacionales.

Una empresa multinacional desea analizar las diferencias salariales en función de dos factores: el **área de trabajo** y el **género del empleado**. Con el interés de examinar si existen diferencias salariales atribuibles al departamento en el que se trabaja, al género de los empleados o a la interacción entre ambos factores, se recopilaron datos de **salarios mensuales** (en dólares) de una muestra representativa de trabajadores en tres áreas clave de la organización: **Manufactura**, **Marketing** e **Investigación y Desarrollo**. En cada departamento se seleccionaron cuatro empleados de género femenino y cuatro empleados de género masculino, obteniendo así un total de 24 observaciones.

Los datos recolectados se detallan a continuación:

- Para cada combinación de área y género, se registraron los salarios individuales.  
- Por ejemplo, en el área de **Manufactura**, los salarios femeninos fueron: 1016, 1007, 875, 968, y los masculinos: 978, 1056, 982, 748.  
- En **Marketing**, los salarios femeninos fueron: 1045, 895, 848, 904 y los masculinos: 1154, 1091, 878, 876.  
- En **Investigación y Desarrollo**, los salarios para mujeres fueron: 770, 733, 844, 771 y para hombres: 926, 1055, 1066, 1088.

**Problema:**  
¿Existe alguna diferencia significativa en los niveles salariales según el área funcional, el género del empleado o la interacción entre ambos factores? Realice un **ANOVA de dos vías** para evaluar los efectos principales de los factores "Área" y "Género", así como su posible interacción, utilizando los salarios como variable dependiente.

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import matplotlib.pyplot as plt

# Estructuración de datos
areas = [
    'Manufactura', 'Manufactura', 'Manufactura', 'Manufactura', 'Manufactura', 'Manufactura', 'Manufactura', 'Manufactura',
    'Marketing', 'Marketing', 'Marketing', 'Marketing', 'Marketing', 'Marketing', 'Marketing', 'Marketing',
    'Investigacion', 'Investigacion', 'Investigacion', 'Investigacion', 'Investigacion', 'Investigacion', 'Investigacion', 'Investigacion'
]

generos = [
    'Femenino', 'Femenino', 'Femenino', 'Femenino', 'Masculino', 'Masculino', 'Masculino', 'Masculino',
    'Femenino', 'Femenino', 'Femenino', 'Femenino', 'Masculino', 'Masculino', 'Masculino', 'Masculino',
    'Femenino', 'Femenino', 'Femenino', 'Femenino', 'Masculino', 'Masculino', 'Masculino', 'Masculino'
]

salarios = [
    1016, 1007, 875, 968, 978, 1056, 982, 748,
    1045, 895, 848, 904, 1154, 1091, 878, 876,
    770, 733, 844, 771, 926, 1055, 1066, 1088
]

data = pd.DataFrame({
    "Area": areas,
    "Genero": generos,
    "Salario": salarios
})

# ANOVA de dos vías con interacción
modelo = ols('Salario ~ C(Area) * C(Genero)', data=data).fit()
anova_results = anova_lm(modelo, typ=2)
print("\n--- ANOVA DE DOS VÍAS ---")
print(anova_results)

# Conclusiones
print("\n--- CONCLUSIONES ---")
if anova_results.loc['C(Area)', 'PR(>F)'] < 0.05:
    print("- Existen diferencias salariales significativas atribuibles al área de trabajo (p < 0.05).")
else:
    print("- No se detectaron diferencias salariales significativas debidas únicamente al área (p ≥ 0.05).")

if anova_results.loc['C(Genero)', 'PR(>F)'] < 0.05:
    print("- Se observa un efecto significativo del género sobre los salarios (p < 0.05).")
else:
    print("- No hay evidencia de una diferencia salarial general por género (p ≥ 0.05).")

if anova_results.loc['C(Area):C(Genero)', 'PR(>F)'] < 0.05:
    print("- Existe una interacción significativa entre área y género (p < 0.05): la diferencia salarial por género depende del área, o viceversa.")
else:
    print("- No se detecta interacción relevante entre área y género (p ≥ 0.05).")

# Prueba post hoc de Tukey sobre todas las combinaciones Área-Género
data["Area_Genero"] = data["Area"] + " / " + data["Genero"]
tukey = pairwise_tukeyhsd(endog=data["Salario"], groups=data["Area_Genero"], alpha=0.05)
print("\n--- PRUEBA POST-HOC DE TUKEY ---")
print(tukey)

# Gráfico: diferencias de medias según Tukey
fig, ax = plt.subplots(figsize=(10, 6))
tukey.plot_simultaneous(ax=ax)
plt.title('Comparaciones múltiples (Tukey HSD) por Área y Género')
plt.xlabel('Salario mensual')
plt.ylabel('Grupo Área / Género')
plt.grid(alpha=0.4)
plt.show()
```

Se recolectaron datos durante cinco meses consecutivos (junio a octubre) para tres variedades: shampoo para cabello **seco**, **normal** y **graso**. El objetivo es determinar si las ventas dependen únicamente del tipo de producto, del mes, o si existe una interacción entre ambos factores.

A continuación se muestran los datos registrados (unidades vendidas por mes y tipo):

| Mes         | Tipo    | Ventas |

**Problema:**

Utilice un **ANOVA de dos vías** para analizar el efecto principal del tipo de producto, del mes y de la interacción tipo × mes sobre el volumen de ventas. Explique si es necesario realizar análisis post-hoc según los resultados obtenidos.

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import matplotlib.pyplot as plt

# 1. Preparar y limpiar los datos
data = {
    'Mes': ['Junio', 'Julio', 'Agosto', 'Septiembre', 'Octubre'] * 3,
    'Tipo': (['Seco']*5) + (['Normal']*5) + (['Graso']*5),
    'Ventas': [
        7, 11, 13, 8, 9,
        9, 12, 11, 9, 10,
        12, 14, 8, 7, 13
    ]
}

df = pd.DataFrame(data)

# Limpiar strings (por si acaso hay espacios, mayúsculas)
df['Mes'] = df['Mes'].str.strip().str.capitalize()
df['Tipo'] = df['Tipo'].str.strip().str.capitalize()

# Verificar valores nulos
print("\n--- VALORES NULOS ---")
print(df.isnull().sum())

# Verificar combinaciones completas
print("\n--- COMBINACIÓN TIPO - MES ---")
print(df.groupby(['Tipo', 'Mes']).size().unstack(fill_value=0))

# 2. ANOVA de dos vías sin interacción (por falta de replicados)
modelo = ols('Ventas ~ C(Tipo) + C(Mes)', data=df).fit()
anova_results = anova_lm(modelo, typ=2)

print("\n--- ANOVA DE DOS VÍAS SIN INTERACCIÓN ---")
print(anova_results)

# 3. Interpretación y conclusiones
print("\n--- CONCLUSIONES ---")
if anova_results.loc['C(Tipo)', 'PR(>F)'] < 0.05:
    print("- Hay diferencias significativas en ventas según el tipo de shampoo (p < 0.05).")
else:
    print("- No se observan diferencias significativas en ventas por tipo de shampoo (p ≥ 0.05).")

if anova_results.loc['C(Mes)', 'PR(>F)'] < 0.05:
    print("- Hay diferencias significativas en ventas según el mes de comercialización (p < 0.05).")
else:
    print("- No se observan diferencias significativas en ventas por mes (p ≥ 0.05).")

# 4. Pruebas post-hoc de Tukey solo si hay diferencias significativas
realizar_posthoc = False
if (anova_results.loc['C(Tipo)', 'PR(>F)'] < 0.05) or (anova_results.loc['C(Mes)', 'PR(>F)'] < 0.05):
    realizar_posthoc = True

if realizar_posthoc:
    if anova_results.loc['C(Tipo)', 'PR(>F)'] < 0.05:
        print("\n--- PRUEBA POST-HOC DE TUKEY POR TIPO DE SHAMPOO ---")
        tukey_tipo = pairwise_tukeyhsd(endog=df['Ventas'], groups=df['Tipo'], alpha=0.05)
        print(tukey_tipo)
        tukey_tipo.plot_simultaneous()
        plt.title("Tukey HSD - Comparación por tipo de shampoo")
        plt.grid(alpha=0.3)
        plt.show()

    if anova_results.loc['C(Mes)', 'PR(>F)'] < 0.05:
        print("\n--- PRUEBA POST-HOC DE TUKEY POR MES ---")
        tukey_mes = pairwise_tukeyhsd(endog=df['Ventas'], groups=df['Mes'], alpha=0.05)
        print(tukey_mes)
        tukey_mes.plot_simultaneous()
        plt.title("Tukey HSD - Comparación por mes")
        plt.grid(alpha=0.3)
        plt.show()
else:
    print("\nNo se requieren pruebas post-hoc porque no hay diferencias estadísticamente significativas por tipo ni por mes (p ≥ 0.05).")
```

```python
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import matplotlib.pyplot as plt

# 1. Preparar los datos
data = {
    'Tamaño': [
        'Chico', 'Chico', 'Chico',
        'Mediano', 'Mediano', 'Mediano',
        'Grande', 'Grande', 'Grande',
        'Chico', 'Chico', 'Chico',
        'Mediano', 'Mediano', 'Mediano',
        'Grande', 'Grande', 'Grande'
    ],
    'Peso': [
        'Pesado', 'Pesado', 'Pesado',
        'Pesado', 'Pesado', 'Pesado',
        'Pesado', 'Pesado', 'Pesado',
        'Ligero', 'Ligero', 'Ligero',
        'Ligero', 'Ligero', 'Ligero',
        'Ligero', 'Ligero', 'Ligero'
    ],
    'Valor': [
        23, 20, 11,
        21, 32, 20,
        25, 26, 20,
        13, 20, 11,
        32, 17, 23,
        17, 15, 8
    ]
}
df = pd.DataFrame(data)

# 2. ANOVA de dos vías con interacción
modelo = ols('Valor ~ C(Tamaño) * C(Peso)', data=df).fit()
anova_results = anova_lm(modelo, typ=2)
print("\n--- ANOVA DE DOS VÍAS ---")
print(anova_results)

# 3. Interpretación y conclusiones
print("\n--- CONCLUSIONES ---")
if anova_results.loc['C(Tamaño)', 'PR(>F)'] < 0.05:
    print("- Hay diferencias significativas en el valor según el tamaño (p < 0.05).")
else:
    print("- No se observan diferencias significativas en el valor por tamaño (p >= 0.05).")
if anova_results.loc['C(Peso)', 'PR(>F)'] < 0.05:
    print("- Hay diferencias significativas en el valor según el peso (p < 0.05).")
else:
    print("- No se observan diferencias significativas en el valor por peso (p >= 0.05).")
if anova_results.loc['C(Tamaño):C(Peso)', 'PR(>F)'] < 0.05:
    print("- Existe interacción significativa entre tamaño y peso (p < 0.05): el efecto del tamaño varía con el peso.")
else:
    print("- No hay interacción significativa entre tamaño y peso (p >= 0.05).")

# 4. Prueba post-hoc de Tukey si corresponde
realizar_posthoc = (
    (anova_results.loc['C(Tamaño)', 'PR(>F)'] < 0.05) or
    (anova_results.loc['C(Peso)', 'PR(>F)'] < 0.05) or
    (anova_results.loc['C(Tamaño):C(Peso)', 'PR(>F)'] < 0.05)
)

if realizar_posthoc:
    print("\n--- PRUEBA POST-HOC DE TUKEY POR TAMAÑO ---")
    tukey_tamano = pairwise_tukeyhsd(endog=df['Valor'], groups=df['Tamaño'], alpha=0.05)
    print(tukey_tamano)
    tukey_tamano.plot_simultaneous()
    plt.title("Tukey HSD - Comparación por Tamaño")
    plt.grid(alpha=0.3)
    plt.show()

    print("\n--- PRUEBA POST-HOC DE TUKEY POR PESO ---")
    tukey_peso = pairwise_tukeyhsd(endog=df['Valor'], groups=df['Peso'], alpha=0.05)
    print(tukey_peso)
    tukey_peso.plot_simultaneous()
    plt.title("Tukey HSD - Comparación por Peso")
    plt.grid(alpha=0.3)
    plt.show()

    if anova_results.loc['C(Tamaño):C(Peso)', 'PR(>F)'] < 0.05:
        print("\n--- PRUEBA POST-HOC DE TUKEY POR COMBINACIÓN TAMAÑO Y PESO ---")
        df['Tamaño_Peso'] = df['Tamaño'] + " / " + df['Peso']
        tukey_inter = pairwise_tukeyhsd(endog=df['Valor'], groups=df['Tamaño_Peso'], alpha=0.05)
        print(tukey_inter)
        tukey_inter.plot_simultaneous()
        plt.title("Tukey HSD - Comparación combinada Tamaño y Peso")
        plt.grid(alpha=0.3)
        plt.show()
else:
    print("\nLas pruebas post-hoc no son necesarias porque no se encontraron diferencias significativas globales (ANOVA p ≥ 0.05).")
```
